### Compute the three pairwise correlation coefficients between each of the three color channels across all pixels:

```python
channel_1 = np.ndarray.flatten(image[:,:,0])
channel_2 = np.ndarray.flatten(image[:,:,1])
channel_3 = np.ndarray.flatten(image[:,:,2])
channel_matrix = [channel_1, channel_2, channel_3]
print(np.shape(channel_matrix))
```

```python
# return Pearson product-moment correlation coefficients between each color channel
correlation_coefficients = np.corrcoef(channel_matrix)
print(correlation_coefficients)
```

### Is there anything striking in the correlations? Why is this not surprising given the way natural images arise?

The correlation is all positive, and each is quite high. This isn't surprising, as the images are taken from a photon receptor in a digital camera (most likely). Therefore, the image is a conglomoration of the 3 color channels together. The image processing in the camere is not very independent of the channel, and thus the final image has a high correlation coefficient between each color channel.

### Once more, plot the three channels separately as gray level images into a single figure together with the color image:

```python
hsv_channel_1 = np.ndarray.flatten(hsv_image[:,:,0])
hsv_channel_2 = np.ndarray.flatten(hsv_image[:,:,1])
hsv_channel_3 = np.ndarray.flatten(hsv_image[:,:,2])
hsv_channel_matrix = [hsv_channel_1, hsv_channel_2, hsv_channel_3]
print(np.shape(hsv_channel_matrix))
```

```python
# return Pearson product-moment correlation coefficients between each color channel
hsv_correlation_coefficients = np.corrcoef(hsv_channel_matrix)
print(hsv_correlation_coefficients)
```

### Explain why there is less correlation now:

The HSV representation is used to make the image more like how the visual system processes an image. Reading about this, the wiki says that the representation models the way paints of different colors mix together, where a saturation element resembles tint and a value measure showing the amount of black or white paint.

It would make sense then that the correlation is much lower, as these channels are not linked any more. Before for the RGB, each channel was a measure of color. Now for HSV, each is a different variable and thus much less correlated. 

### #3. Decompose the image into the separate HSV-channels: 

```python
hsv_channels = [hsv_image[:,:,0], hsv_image[:,:,1], hsv_image[:,:,2]]
```

### Perform the following computations for each channel separately: 
### For each pixel in the image, extract the surrounding 15 × 15 pixel image patch.
### For each patch compute the standard deviation of the pixel values.

```python
pixels = [[],[],[]]
for i in range(len(hsv_channels)):
    height = np.shape(hsv_channels[i])[0]  # height of the image
    width = np.shape(hsv_channels[i])[1]  # width of the image
    size = 7  # size of half the 15x15 patch, not including the center pixel
    pixels[i] = np.zeros((height-size*2, width-size*2))  # empty pixel array
    for j in range(size, height-size):
        for k in range(size, width-size):
            # compute the standard deviation at each pixel for the patch size
            samples = hsv_channels[i][j-size:j+size+1, k-size:k+size+1]  # create the patch
            pixels[i][j-size,k-size] = circstd(samples=samples, high=1)  # calculate the std
print(np.shape(pixels))
```

### Store the standard deviation of the patches around each pixel in a new “saliency image”.

```python
saliency_image = pixels
```

### #4. Plot the separate saliency images into a single figure. 

```python
fig, axs = plt.subplots(1, 3)
axs[0].imshow(saliency_image[0], cmap='Greys')
axs[1].imshow(saliency_image[1], cmap='Greys')
axs[2].imshow(saliency_image[2], cmap='Greys')
plt.show()
```

### How is each of the three images related to the visual properties of the image?

Now that we are in the HSV space, each image above shows the saliency with respect to the hue (radial slice of color), saturation (tint), and value (amount of black or white) of the original image. 

### #5. Compute a joint saliency map by summing the separate saliency images you have computed.

```python
joint_saliency_map = np.sum(saliency_image, axis=0)
print(np.shape(joint_saliency_map))
```

### Add this plot to the previous figure.

```python
ax1 = plt.subplot(212)
ax2 = plt.subplot(231)
ax3 = plt.subplot(232)
ax4 = plt.subplot(233)
ax1.imshow(joint_saliency_map, cmap='Greys')
ax1.set_title('Joint Saliency Map')
ax2.imshow(saliency_image[0], cmap='Greys')
ax3.imshow(saliency_image[1], cmap='Greys')
ax4.imshow(saliency_image[2], cmap='Greys')
plt.show()
```

### #6. Smooth this image by low-pass filtering. Experiment with different sigmas and padding modes

```python
sigmas = [1, 2, 5]
modes = ['reflect', 'mirror', 'constant']
fig, axs = plt.subplots(3, 3)
for i in range(3):
    for j in range(3):
        smoothed_saliency = gaussian_filter(input=joint_saliency_map, sigma=sigmas[i], mode=modes[j])
        axs[i,j].imshow(smoothed_saliency, cmap='Greys')
        axs[i,j].set_ylabel(r'sigma = %i' %sigmas[i])
        axs[i,j].set_title(r'mode = %s' %modes[j])
plt.tight_layout()
plt.show()
```

The plots above show testing different sigma and mode values for the gaussian filter function.

### Plot the result into a new figure.

```python
smoothed_saliency = gaussian_filter(input=joint_saliency_map, sigma=3, mode='constant')
plt.imshow(smoothed_saliency, cmap='Greys')
plt.show()
```

### #7. In a new figure, plot the original image and overlay the contours of the smoothed saliency map

```python
whole_contour = np.zeros((height, width))  # empty corrected saliency map
print(np.shape(whole_contour), np.shape(smoothed_saliency))
for i in range(7, height-7):
    for j in range(7, width-7):
        whole_contour[i,j] = smoothed_saliency[i-7,j-7]
```

```python
plt.contour(whole_contour)
plt.imshow(image, alpha=0.5)
plt.show()
```

### #8. From the smoothed saliency map, extract the positions the subject will most likely and least likely look at.

```python
max_index = np.unravel_index(np.argmax(whole_contour), np.shape(whole_contour))
print(max_index)
min_index = np.unravel_index(np.argmin(whole_contour), np.shape(whole_contour))
print(min_index)
```

### Mark those points with a red and blue dot, respectively.

```python
plt.contour(whole_contour)
plt.imshow(image, alpha=0.5)
plt.plot(max_index[1], max_index[0], marker='o', color='r')
plt.plot(min_index[1], min_index[0], marker='o', color='b')
plt.show()
```

## Exercise 2: Mexican-hat convolution (optional/bonus task)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import rgb_to_hsv
from scipy.stats import circstd
from scipy.signal import fftconvolve, convolve2d
from helper import  mexican_hat
from os import listdir
%matplotlib inline
```

## Mexican hat

```python
# get the image 
img = plt.imread("salience_03.png")
xsize, ysize, channels = img.shape
# build the kernel
mex_kernel = mexican_hat(xsize, ysize, width=3)
# get hsv image
img_hsv = rgb_to_hsv(img)
# Convolve the image with the prepared “mexican hat” filter
sal_map = np.zeros_like(img_hsv)
for i in range(3): 
    sal_map[:,:,i] = np.abs(convolve2d(img_hsv[:,:,i], mex_kernel, mode='same', boundary='symm'))
```

```python
plt.figure(figsize=(15,10))
strings = ["hue", "saturation", "value"]
for i in range(3): 
    plt.subplot(2,2,i+1)
    plt.imshow(sal_map[:,:,i], cmap='gray')
    plt.title("Saliency image from the " + strings[i] + " channel")   
```
