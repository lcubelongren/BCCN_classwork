```python
def spike_or_not(data):
    time_range = np.linspace(0,4,40001)  # Create an array to put all the spike times into
    count = -1  # Start at -1, as we will initially add one and bring it to zero
    for i in time_range:  # Checks if the value exists in the raw spike train file
        count += 1
        if i in data:  # If the time value shows up in the data, make it to a one
            time_range[count] = 1
        else: 
            time_range[count] = 0  # Otherwise, make it a zero
    return time_range

binary_spikes = spike_or_not(spiketimes1_seconds)
# This gives us a list of every time point, where a spike is a 1 and no spike is a 0
```

#### c) Use the window functions defined below to estimate an approximate firing rate, where $Ï(t)$ is the neural response function and T is the trial length.

$ r_{approx}(t) = \sum_{i=1}^{n} w \ (t âˆ’t_i) = \int_{0}^{T} d\tau \ w(t - \tau) \ Ï(\tau) $

```python
def firing_rate_plot(function, modifier, bin_size):
    x = np.linspace(0, 4, len(function))  # Set x-axis manually in order to control scale
    normalized_function = function/(bin_size/1000) # Normalize to get value of spikes/s
    plt.plot(x, normalized_function, color='k') 
    plt.title('Firing Rate w/ '+str(modifier)+' = '+str(bin_size)+'ms')
    plt.ylabel('spikes per second')
    plt.xlabel('time [s]')
    plt.grid()
    plt.show()
```

```python
# Let's define tau here. From what I can see, it is the domain. 
# We shall say that it is -T/2 -> T/2
# Then we will be able to calculate the kernel

T = 4000 # Trial length in ms
domain = np.linspace(-T/2, T/2, len(binary_spikes))
```

#### i) Estimate the rate with a rectangular window with widths of $âˆ†t$ = 20, 50 and 150 ms. The rectangular window is centered at each spike (non causal filtering).

$ w(\tau)=   \left\{
\begin{array}{ll}
      1/ \Delta t & -\Delta t/2 \leq \tau \lt \Delta t/2 \\
      0 & otherwise\\
\end{array} 
\right.  $


```python
def rectangular_window(tau, dt):
    w = np.zeros(len(tau))
    for i in range(len(tau)):  # Writing out the piecewise function
        if -dt/2 <= tau[i] < dt/2:
            w[i] = 1/dt
        else:
            w[i] = 0
    return w
```

```python
def rectangular_convolution(tau, data, bin_size):  # Here is the convolution, between the window and the data
    return np.convolve(rectangular_window(tau, bin_size), data, mode='same')

firing_rate_plot(function=rectangular_convolution(domain, binary_spikes, 20), modifier='$\Delta t$', bin_size=20)
firing_rate_plot(function=rectangular_convolution(domain, binary_spikes, 50), modifier='$\Delta t$', bin_size=50)
firing_rate_plot(function=rectangular_convolution(domain, binary_spikes, 150), modifier='$\Delta t$', bin_size=150)
```

#### ii) Estimate the rate with a Gaussian window with $Ïƒ_w$ = 10, 20 and 50 ms. Here the Gaussian is centered at each spike (non causal filtering).

$ w(\tau) = \frac{1}{\sqrt{2 \pi}\sigma_w} exp \left(- \frac{\tau^2}{2\sigma_w^2} \right) $

```python
def gaussian_window(tau, sigma):  # Writing the Gaussian window function
    w = (1 / (np.sqrt(2*np.pi) * sigma)) * np.exp(-tau**2 / (2*sigma**2))
    return w
```

```python
def gaussian_convolution(tau, data, sigma):  # Convolution again
    return np.convolve(gaussian_window(tau, sigma), data, mode='same')

firing_rate_plot(function=gaussian_convolution(domain, binary_spikes, 10), modifier='$\sigma_w$', bin_size=10)
firing_rate_plot(function=gaussian_convolution(domain, binary_spikes, 20), modifier='$\sigma_w$', bin_size=20)
firing_rate_plot(function=gaussian_convolution(domain, binary_spikes, 50), modifier='$\sigma_w$', bin_size=50)
```

#### iii)  Estimate the rate with an alpha function where $1/Î±$ is a time scale and $[x]_+$ is the half-wave rectification function. Use $1/Î±$ = 10, 20 and 50 ms.
#### Here the rate starts rising only after a spike is emitted, i.e., the filter is causal. Why would one want to use a causal rather than a non-causal filter?


$ w(\tau) = \left[ \alpha^2 \ \tau \ exp \ (-\alpha \tau ) \right]_+ $

```python
def alpha_function(tau, alpha_invert):  # Writing the alpha function
    alpha = 1 / alpha_invert
    arg = (alpha**2) * tau * np.exp(-alpha*tau)
    w = np.clip(arg, 0, None)  # Perform the half-wave rectification
    return w  # Any negative values are set to zero
```

```python
def alpha_convolution(tau, data, alpha_invert):  # Again the convolution
    return np.convolve(alpha_function(tau, alpha_invert), data, mode='same')

firing_rate_plot(function=alpha_convolution(domain, binary_spikes, 10), modifier=r'$\alpha^{-1}$', bin_size=10)
firing_rate_plot(function=alpha_convolution(domain, binary_spikes, 20), modifier=r'$\alpha^{-1}$', bin_size=20)
firing_rate_plot(function=alpha_convolution(domain, binary_spikes, 50), modifier=r'$\alpha^{-1}$', bin_size=50)
```

One would want to use a causal filter, rather than a non-causal filter, due to this being closer to the biological process that occurs for neurons. From section 1.2 of D&A, "[a] postsynaptic neuron monitoring the spike train of a presynaptic cell has access only to spikes that have previously occurred." Therefore, using the alpha function to determine firing rate is the most accurate method when trying to simulate the biological aspects of these processes.

#### d) Calculate the spike count-rate:

$ r = \frac{n}{T} = \frac{1}{T} \int_0^T dt \ p(t) $ 

```python
def spike_count_rate(data, period):  # Function to return the spike count-rate
        n = len(data)  # Spike count
        T = period  # Period
        rate = n / T  # Simple division to determine rate
        return rate
```

```python
print('Spike count-rate:', spike_count_rate(data=spiketimes1, period=T/1000), 'spikes/s')
```

### Nicer Plotting:

```python
# Still trying to learn the best method for subplots
# Here I plot together all of what was done before, all for the largest modifier size
plt.rcParams["figure.figsize"] = (16,8)
def firing_rate_subplots(function0, function1, function2, function3, modifier1, modifier2, modifier3, bin_size):
    x = np.linspace(0, 4, len(function1))  # Set x-axis manually in order to control scale
    normalized_function1 = function1/(bin_size/1000) # Normalize to get value of spikes/s
    normalized_function2 = function2/(bin_size/3000)
    normalized_function3 = function3/(bin_size/3000)
    
    bin_size_hist = 150
    counts, bins = np.histogram(function0, 4000//bin_size_hist)  # Create the histogram
    normalized_counts = counts/(bin_size_hist/1000)
    
    fig, axes = plt.subplots(5)
    axes[0].eventplot(spiketimes1_seconds, color='k', label='Raw Spike Train')
    axes[0].legend()
    axes[1].hist(bins[:-1], bins, weights=normalized_counts, histtype='step', color='k', label=r'Histogram $\Delta t$ = 150ms')
    axes[1].legend()
    axes[2].plot(x, normalized_function1, color='k', label=r'Rectangular Window $\Delta t$ = 150ms')
    axes[2].legend()
    axes[3].plot(x, normalized_function2, color='k', label=r'Gaussian Window $\sigma_w$ = 50ms')
    axes[3].legend()
    axes[4].plot(x, normalized_function3, color='k', label=r'Alpha Function $\alpha^{-1}$ = 50ms')
    axes[4].legend()
    
    fig.tight_layout()
    axes[0].set_title('Firing Rates')
    axes[0].set_ylabel('spikes/s')
    axes[1].set_ylabel('spikes/s')
    axes[2].set_ylabel('spikes/s')
    axes[3].set_ylabel('spikes/s')
    axes[4].set_ylabel('spikes/s')
    plt.xlabel('time [s]')
    plt.show()
```

```python
bin_size_rectangle = 150  # [ms]
bin_size_gaussian = 50  # [ms]
bin_size_alpha = 50  # [ms]
firing_rate_subplots(function0=spiketimes1_seconds, function1=rectangular_convolution(domain, binary_spikes, bin_size_rectangle), 
                     function2=gaussian_convolution(domain, binary_spikes, bin_size_gaussian), function3=alpha_convolution(domain, binary_spikes, bin_size_alpha)
                     , modifier1='$\Delta t$', modifier2='$\sigma_w$', modifier3=r'$\alpha^{-1}$', bin_size=bin_size_rectangle)
```

Above is a collection of the plots, plotted in a similar way as seen in Dayan & Abbott. Here I chose to use the largest given bin sizes or modifiers in order to easily see differences between the windows/functions.
