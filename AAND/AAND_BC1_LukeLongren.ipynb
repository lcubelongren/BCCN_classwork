## Load the data set

```python
fname = 'eyes_closed_VPal.npz'
X, fs, clab, mnt = bci.load_data(fname)
"""
X: EEG signals (channels, time)
fs: sampling rate of the signals
clab: labels of the corresponding channels
mnt: channel coordinates
"""
print(np.shape(X), ', shape of the EEG data')
```

## Exercise 1: PCA on raw data  (2 points)
Make a scatter plot using the channels Fz and Oz. Visualize the complete covariance matrix (functions `np.cov` and `plt.imshow`). Then determine the two principle components (Eigenvectors) (function `np.linalg.eigh`) of that data that correspond to the largest Eigenvalues and display them as scalp maps (function `scalpmap` provided in the `bci_minitoolbox`).

```python
# scatter plot of the Fz, Oz channels
Fz_idx, Oz_idx = clab.index('Fz'), clab.index('Oz')
plt.scatter(X[Fz_idx], X[Oz_idx], c=np.arange(0, np.shape(X)[1]/fs, 1/fs))
plt.title('Raw Data - Channels Fz and Oz')
plt.xlabel('Fz [$\mu$V]'), plt.ylabel('Oz [$\mu$V]')
plt.colorbar(label='time [s]')
plt.grid()
plt.show()
```

```python
def PCA(data):
    cov = np.cov(data, rowvar=True)  # measure covariances, construct the matrix
    eigenvalues, eigenvectors = LA.eigh(cov)  # eigenvalue decomposition
    return cov, eigenvectors, eigenvalues
```

```python
# center the data
X_centered = X - np.mean(X)
cov, PCs, eigenvalues = PCA(X_centered)
```

```python
# covariance matrix
plt.imshow(cov, origin='lower')
plt.title('Covariance Matrix of the Data')
plt.xlabel('channel index'), plt.ylabel('channel index')
plt.colorbar(label='covariance')
plt.show()
```

```python
# smap_limits = [-0.1, 0.2]  # colorbar range
smap_limits = 'minmax'

plt.figure()
plt.subplot(1,2,1)
bci.scalpmap(mnt=mnt, v=PCs[:, -1:], clim=smap_limits, cb_label='principle component')
plt.title('Largest Eigenvalue')

plt.subplot(1,2,2)
bci.scalpmap(mnt=mnt, v=PCs[:, -2:-1], clim=smap_limits, cb_label='principle component')
plt.title('Second Largest Eigenvalue')
plt.show()
```

## Exercise 2: Artifact Correction with PCA  (3 points)
Using the backward model, estimate the source activity corresponding to the two components that were found in Ex. \#01 and plot the two obtained time series. Using the forward model, estimate that part of the measured EEG signals, that originates from these two sources. Subtract this multivariate signal from the original EEG. Compare the thus obtained cleaned signal in channel Fz with the original data.

## <font color='k'> Backward Model: $\hat{s}(t) = W^T x(t)$

where $s$ is the current source, $W$ the pseudoinverse of the distribution vectors, and $x$ are scalp potentials.

```python
source_activity = [LA.pinv(PCs[:, -1:]) @ X,  # W.T*x(t) for two eigenvalues
                   LA.pinv(PCs[:, -2:-1]) @ X]  
time_range = np.arange(0, np.shape(X)[1]/fs, 1/fs)  # plotting time

fig, axs = plt.subplots(np.shape(source_activity)[0], 1, sharex=True, sharey=True, figsize=(12,8))
for j in range(np.shape(source_activity)[0]):  # loop over eigenvalues
    axs[j].plot(time_range, source_activity[j].T, color='k', label='#%i Largest Eigenvalue' %(j+1))
    axs[j].set_ylabel('potential  [$\mu$V]')
    axs[j].legend()
    axs[j].grid()
axs[0].set_title('Estimated Source Activity')
axs[1].set_xlabel('time [s]')
plt.tight_layout()
plt.show()
```

## <font color='k'> Forward Model: $x(t) = A \ s(t) + n(t)$

where $x$ is the scalp potential, $A$ is composed of distribution vectors, $s$ is the current source, and $n$ is the noise.

Here, we will just take the original data $X$ and subtract the two eigenvalue components dotted with the estimated activity:

```python
# estimate that part of the measured EEG signals originating from these two sources
# clean the data by subtracting this multivariate signal from the original EEG
X_cleaned = X - np.dot(PCs[:, -1:], source_activity[0]) - np.dot(PCs[:, -2:-1], source_activity[1])
print(np.shape(X_cleaned))
```

```python
fig, axs = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(12,8))
axs[0].plot(time_range, X[Fz_idx], color='k', label='original signal')
axs[1].plot(time_range, X_cleaned[Fz_idx], color='k', label='cleaned signal')
axs[0].set_ylabel('potential  [$\mu$V]'), axs[1].set_ylabel('potential  [$\mu$V]')
axs[0].legend(), axs[1].legend()
axs[0].grid(), axs[1].grid()
axs[0].set_title('Original and Cleaned Signal - Channel Fz')
axs[1].set_xlabel('time [s]')
plt.tight_layout()
plt.show()
```

## Exercise 3: PCA on band-pass filtered data (3 points)
Design a butterworth band-pass filter of order 10 with the frequency band 8-12 Hz (function `sp.signal.butter`). Apply that band-pass filter to the original EEG signals (function `sp.signal.lfilter`) and compare raw and filtered signals for channel Oz.  
Visualize the covariance matrix of the filtered data and check the eigenvalue spectrum, to determine how many strong principal components there are. Visualize those principal components as in Ex. \#01. Compare the results and discuss shortly.

```python
# butterworth band-pass filter
b, a = sp.signal.butter(N=10, Wn=[8,12], btype='bandpass', fs=fs)  # numerator, denominator coefficient 
w, h = sp.signal.freqs(b, a)  # generate a bunch of filter values
plt.semilogx(w, 20 * np.log10(abs(h)), color='k')  # plot the response curve of the created butterworth filter
plt.title('Butterworth Band-Pass Filter Frequency Response')
plt.xlabel('frequency [radians / second]')
plt.ylabel('amplitude')
plt.grid(which='both')
plt.show()
```

```python
# apply the band-pass filter to the original EEG signals
X_filtered = sp.signal.lfilter(b=b, a=a, x=X)

fig, axs = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(12,8))
axs[0].plot(time_range, X[Oz_idx], color='k', label='original signal')
axs[1].plot(time_range, X_filtered[Oz_idx], color='k', label='filtered signal')
axs[0].set_ylabel('potential  [$\mu$V]'), axs[1].set_ylabel('potential  [$\mu$V]')
axs[0].legend(), axs[1].legend()
axs[0].grid(), axs[1].grid()
axs[0].set_title('Original and Filtered Signal - Channel Oz')
axs[1].set_xlabel('time [s]')
plt.tight_layout()
plt.show()
```

```python
# covariance matrix of filtered data
cov_filtered, PCs_filtered, eigenvalues_filtered = PCA(X_filtered)
plt.imshow(cov_filtered, origin='lower')
plt.title('Covariance Matrix of Filtered Data')
plt.xlabel('channel index'), plt.ylabel('channel index')
plt.colorbar(label='covariance')
plt.show()
```

```python
chan_range = np.linspace(1, np.shape(X)[0], np.shape(X)[0])  # channel number array
color_choice = [[],[]]  # original, filtered color for scatter
for i in range(np.shape(chan_range)[0]):
    if eigenvalues[i] > 1e4:
        color_choice[0].append('r')
    else:
        color_choice[0].append('k')
    if eigenvalues_filtered[i] > 1e3:
        color_choice[1].append('r')
    else:
        color_choice[1].append('k')

fig, axs = plt.subplots(2, 1, sharex=True, sharey=False, figsize=(12,8))
axs[0].scatter(chan_range, eigenvalues, label='original signal', c=color_choice[0])
axs[1].scatter(chan_range, eigenvalues_filtered, label='filtered signal', c=color_choice[1])
axs[0].set_ylabel('eigenvalue'), axs[1].set_ylabel('eigenvalue')
axs[0].legend(), axs[1].legend()
axs[0].grid(), axs[1].grid()
axs[0].set_yscale('log'), axs[1].set_yscale('log')
axs[0].set_title('Original and Filtered Eigenvalues')
axs[1].set_xlabel('channel #')
plt.tight_layout()
plt.show()
```

Inspecting the eigenvalues of the original and filtered data, we can see that while the original data had two outlier/dominant eigenvalues, the filtered data has three.  
These particular eigenvalues are shown in red in the above graphs.

```python
smap_limits = 'minmax'

plt.figure()
plt.subplot(1,3,1)
bci.scalpmap(mnt=mnt, v=PCs_filtered[:, -1:], clim=smap_limits, cb_label='principle component')
plt.title('Largest Eigenvalue')

plt.subplot(1,3,2)
bci.scalpmap(mnt=mnt, v=PCs_filtered[:, -2:-1], clim=smap_limits, cb_label='principle component')
plt.title('Second Largest Eigenvalue')

plt.subplot(1,3,3)
bci.scalpmap(mnt=mnt, v=PCs_filtered[:, -3:-2], clim=smap_limits, cb_label='principle component')
plt.title('Third Largest Eigenvalue')
plt.show()
```

Discussion of these results:  
Looking at the covariance matrices of the filtered data, the peak covariance values are at the opposite corner than that of the original data.  
The magnitude of the eigenvalues for the filtered data are also smaller than the original data.  
Comparing the scalp maps of the two data inputs, the filtered scalp map looks like a flipped version of the original scalp map.

## Exercise 4: Reconstruction with few PCs  (2 points)
Calculate the signals corresponding to the strongest principal components (second part of Ex. \#03) of the filtered data (project EEG data onto the PCs, i.e. go from *sensor space* to *source space*). Then reconstruct the EEG data from those strong principal components, i.e. project the signals obtained in the first step back into *sensor space*. Compare the band-pass filtered signal with the reconstructed signal in channel Oz. For plotting, the first six seconds are enough. You can also apply an error measure on your choice.

```python
# sensor space to source space
source_activity_filtered = [LA.pinv(PCs_filtered[:, -1:]) @ X_filtered, 
                            LA.pinv(PCs_filtered[:, -2:-1]) @ X_filtered, 
                            LA.pinv(PCs_filtered[:, -3:-2]) @ X_filtered]
reconstructed_data = (np.dot(PCs_filtered[:, -1:], source_activity_filtered[0]) 
                      - np.dot(PCs_filtered[:, -2:-1], source_activity_filtered[1]) 
                      - np.dot(PCs_filtered[:, -3:-2], source_activity_filtered[2]))

time_range = np.arange(0, np.shape(X)[1]/fs, 1/fs)  # plotting time
plt.plot(time_range, reconstructed_data[Oz_idx], color='k', label='reconstructed data', linestyle='-')
plt.plot(time_range, X_filtered[Oz_idx], color='b', label='filtered data', linestyle='--')
plt.title('Reconstructed and Filtered Activity - Channel Oz')
plt.ylabel('potential  [$\mu$V]')
plt.xlabel('time [s]')
plt.xlim(0, 6)  # only the first six seconds
plt.legend()
plt.grid()
plt.show()
```

```python
# error measure - mean squared error
def MSE(data1, data2):
    return (data1 - data2) ** 2 * (1/np.shape(data1)[0])  
```

```python
error_measure = MSE(reconstructed_data[Oz_idx], X_filtered[Oz_idx])

plt.plot(time_range, error_measure, color='k', label='MSE', linestyle='-')
plt.title('Error between Reconstruction and Filtered Signal - Channel Oz')
plt.ylabel('error measure')
plt.xlabel('time [s]')
plt.xlim(0, 6)  # only the first six seconds
plt.legend()
plt.grid()
plt.show()
```
