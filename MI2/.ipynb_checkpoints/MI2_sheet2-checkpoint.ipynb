Center the data first:

```python
def PCA(data):
    cov = np.cov(data.T)
    w, v = LA.eig(cov)
    PC = data @ v
    return PC, w, v
```

## 2.2: Whitening

```python
data_pca4 = np.loadtxt(open(r"C:\Users\lcube\Desktop\jupyter\BCCN\MI2\given\sheet2\pca4.csv", "rb"), delimiter=",", skiprows=1)
print('Shape of data:', np.shape(data_pca4))
```

```python
fig, axs = plt.subplots(1,4)
axs[0].plot(data_pca4[:,0], color='k')
axs[0].set_xlabel('$x_1$')
axs[1].plot(data_pca4[:,1], color='k')
axs[1].set_xlabel('$x_2$')
axs[2].plot(data_pca4[:,2], color='k')
axs[2].set_xlabel('$x_3$')
axs[3].plot(data_pca4[:,3], color='k')
axs[3].set_xlabel('$x_4$')
plt.tight_layout()
plt.show()
```

Briefly looking at the data visually, we see $x_3$ and $x_4$ have a couple outliers that are greater than +-10.
Therefore, we can simply remove values not between this range.

```python
def data_cleaner(data):
    size = len(data[0])  # number of variables
    points = len(data[:, 0])  # number of points per variable
    to_delete = []
    for i in range(size):  # over each variable
        for j in range(points):  # over every point
            if data[j, i] > 10:
                to_delete = np.append(to_delete, j)
            elif data[j, i] < -10: 
                to_delete = np.append(to_delete, j)
            else:
                pass
    cleaned_data = np.delete(data, to_delete, axis=0)
    return cleaned_data
```

```python
cleaned_data_pca4 = data_cleaner(data_pca4)
print(np.shape(data_pca4))
print(np.shape(cleaned_data_pca4))
```

Now we have cleaned data.

### (c) â€œWhitenâ€ the data, i.e. transform the data into 4 new uncorrelated-unit-variance variables. 
### Each new variable should have mean 0 and a standard deviation equal to 1. 
### This can be done e.g. using the transformation:  $V = Î›^{âˆ’1/2} M^{\top} X$ (or $V = M^{\top} X M Î›^{âˆ’1/2}$)

```python
def whitening(data):
    # center the data
    mean = np.mean(data, axis=0)  # Returns center point (x,y)
    centered_data = data - mean
    
    cov = np.cov(data.T)  # compute covarience matrix of centered data
    w, v = LA.eig(cov)  # compute eigens of covarience matrix

    diag_w = np.zeros((len(w), len(w)))  # convert eigenvalues to diagonal matrix
    for i in range(len(w)):
        diag_w[i,i] = w[i]
    
    # calculate new variables
    '''
    the variables here can be a bit confusing. let's define them here:
    M -> normalized eigenvectors of covarience matrix (v)
    X -> centered data (centered_data)
    Î› -> diagonal matrix of eigenvalues (diag_w)
    V -> new variables (whitened_variables)
    '''
    whitened_variables = ((centered_data @ v.T) @ v) @ LA.inv(diag_w)**0.5
    return whitened_variables,centered_data,cov,w,v
```

```python
whitened_data_pca4 = whitening(reasonable_data_pca4)[0]
print(np.shape(whitened_data_pca4))
```

## 2.3: Ojaâ€™s Rule: Derivation

### Derive an approximation to this update rule for a small value of the learning rate parameter $Îµ$ by Taylor-expanding the right hand side of the equation (*) with respect to $Îµ$ around $Îµ = 0$. Show that neglecting terms of second or higher order in Îµ gives Ojaâ€™s rule.

## <font color='black'> $ w_i (t+1) = \frac{w_i(t) + \epsilon y(t) x_i(t)}{(\sum_{j=1}^N [w_j(t) + \epsilon y(t) x_j(t)]^2)^{1/2}} \ \ (*) $

We shall perform a Taylor expansion of the equation (*) in order to derive a more simple approximation.

In the general case, this is as follows:

## <font color='black'> $ f(x) = f(a) + \frac{f'(a)}{1!}(x-a) + \frac{f''(a)}{2!}(x-a)^2 + ...,$ around $a$

Applying the expansion to our equation, around $Îµ = 0$, to only the first order we recieve:

## <font color='black'> $ w_i(t+1) = \frac{w_i(\epsilon) + \epsilon y(\epsilon) x_i(\epsilon)}{(\sum_{j=1}^N [w_j(\epsilon) + \epsilon y(\epsilon) x_j(\epsilon)]^2)^{1/2}} + w_i'(\epsilon) \frac{w_i(\epsilon) + \epsilon y(\epsilon) x_i(\epsilon)}{(\sum_{j=1}^N [w_j(t) + \epsilon y(\epsilon) x_j(\epsilon)]^2)^{1/2}} + ...$

## <font color='black'> $ = \frac{w_i(0)}{(\sum_{j=1}^N [w_j(0)])} + w_i'(0) \frac{w_i(t)}{(\sum_{j=1}^N [w_j(t)])} + ...$

## <font color='black'> $ = \frac{w_i(0)}{(\sum_{j=1}^N [w_j(0)])} + [\frac{w_i(0)}{(\sum_{j=1}^N [w_j(0)])}]' \frac{w_i(t)}{(\sum_{j=1}^N [w_j(t)]} + ...$

## <font color='black'> $ = \frac{w_i}{(\sum_{j=1}^N [w_j])} + \epsilon \ (\frac{y(t) x_i(t)}{(\sum_{j=1}^N [w_j])} - \frac{w_i(t) \sum_{j=1}^N y(t) x_j(t) w_j(t)}{(\sum_{j=1}^N [w_j^{2}])} ) + ...$

Here we shall ensure that the weights are normalized to 1, necessary for stability of the system, by: 

## <font color='black'> $|w| = \sum_{j=1}^N w_j = 1 $

Thus, we result in:

## <font color='black'> $ w_i(t+1) = w_i(t) + \epsilon \ y(t) [x_i(t) - y(t) w_i(t)] $

## 2.4: Ojaâ€™s Rule: Application

### The file data-onlinePCA.txt contains observations from an artificial experiment run over an interval of time 
### (i.e. the first datapoint was observed at $t_0 = 0$ and the last at $t_T = 10s$)

```python
# np.loadtxt(r'C:\Users\lcube\Desktop\jupyter\BCCN\MI2\given\sheet2\data-onlinePCA.txt')
data_onlinePCA = np.loadtxt(r'C:\Users\lcube\Desktop\jupyter\BCCN\MI2\given\sheet2\data-onlinePCA.txt', delimiter=',', skiprows=1, usecols=[1,2])
print(np.shape(data_onlinePCA))  # data of the form [V1,V2]
```

### (a) Produce a scatter plot of the data and indicate the time index by the color of the datapoints:

```python
def plot_data_ojas(data, period, block_number, color_map):
    T = period  # period [s]
    blocks = block_number  # number of blocks
    points = len(data_onlinePCA)  # total data point number
    time_range = points // blocks

    color = iter(color_map((np.linspace(0, 1, blocks))))
    #color = iter(cm.rainbow(np.linspace(0, 1, blocks)))
    for i in range(blocks):
        c = next(color)
        # below, for each iteration (one for each block) we plot the data in the blocked time
        bottom_bound, top_bound = i*time_range, i*time_range + time_range
        plt.scatter(data_onlinePCA[bottom_bound:top_bound, 0], data_onlinePCA[bottom_bound:top_bound, 1], color=c)
    
    plt.title('The Data')
    plt.xlabel('$V_1$')
    plt.ylabel('$V_2$')
    plt.grid()
    
    cmap = color_map
    norm = cm.colors.Normalize(vmin=0, vmax=10)
    cbar = plt.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), boundaries=np.linspace(0,period,period+1))
    cbar.ax.set_ylabel('Time Index [s]')
```

```python
plot_data_ojas(data=data_onlinePCA, period=10, block_number=10, color_map=cm.magma)
plt.show()
```

### (b) Determine the principal components (using batch PCA) for each of the 10 blocks separately.
### Plot the first PC of each block (e.g. as an arrow or the endpoint of it) together with the original data:

```python
def batch_PCA(data, blocks):
    batch_PCA = {}
    time_range = len(data) // blocks
    for i in range(blocks):
        bottom_bound, top_bound = i*time_range, i*time_range + time_range
        batch_PCA[i] = PCA(data[bottom_bound:top_bound])[0]
    return batch_PCA
```

```python
# calculate all the PCs
blocks = 10
batch_PCs = batch_PCA(data=data_onlinePCA, blocks=blocks)

# plot the data with the first PC of each block
cmap = cm.viridis
norm = cm.colors.Normalize(vmin=1, vmax=10)
plot_data_ojas(data=data_onlinePCA, period=10, block_number=10, color_map=cm.Greys)
color = iter(cmap((np.linspace(0, 1, 10))))
for i in range(10):  # for each block
    c = next(color)
    first_PC = batch_PCs[i][0]
    plt.arrow(0, 0, first_PC[0], first_PC[1], width=0.05, color=c)
    
cbar = plt.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), boundaries=np.linspace(0,blocks,blocks+1))
plt.title('Principal Components over the Data')
plt.show()
```

### (c) Implement Ojaâ€™s rule and apply it with a learning rate parameter Îµ âˆˆ {0.002, 0.04, 0.45} to the dataset.
### In each iteration, take the next data point and apply the learning step. 
### Plot the weights at each timestep in the same plot as the original data.
### Interpret your results:

```python
def ojas_rule(learning_rate, data, blocks, iterations):
    w = np.array([np.zeros((blocks, iterations)), np.zeros((blocks, iterations))])  # dimensions (2,blocks,iterations)
    y = np.zeros(iterations)
    w[0][:,0], w[1][:,0] = 1, 1  # set initial weights
    for t in range(iterations-1):
        for i in range(blocks):
            weight = np.array([w[0][i,t], w[1][i,t]])  # the weight found from previous iterations
            x = data[i]  # dimensions (2,1) the next data point
            y[t] = x @ weight  # dimensions (1)
            new_weight = weight + learning_rate * y[t] * (x - y[t]*weight)  # apply the learning step
            w[0][i,t+1], w[1][i,t+1] = new_weight[0], new_weight[1] 
    return w[0], w[1]
```

```python
epsilon = [0.002, 0.04, 0.45]  # the learning rates
iteration_num = int(100)  # set number of iterations

for e in epsilon:  # for each learning rate
    # plot the original data
    plot_data_ojas(data=data_onlinePCA, period=10, block_number=10, color_map=cm.Greys)
    # plot the data after implementing Oja's rule, plotting weights as (w1,w2)
    cmap = cm.viridis  # choose color values
    norm = cm.colors.Normalize(vmin=1, vmax=iteration_num)  # set colorbar range
    color = iter(cmap((np.linspace(0, 1, 10))))
    for i in range(10):  # for each block
        c = next(color)
        ojas = ojas_rule(learning_rate=e, data=data_onlinePCA, blocks=10, iterations=iteration_num)
        plt.scatter(ojas[0][:,i], ojas[1][:,i], color=c, marker='x', s=100)
    plt.title('Weights over the data for $\epsilon = %s$' %e)
    cbar = plt.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap))
    cbar.ax.set_ylabel('Iteration #')
    plt.show()
```

Inspecting our result, we can see that increasing the learning rate results in a quicker change in value of the weights over the iterations.
