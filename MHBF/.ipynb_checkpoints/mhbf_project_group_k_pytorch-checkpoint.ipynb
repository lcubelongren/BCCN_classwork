```python
# Import the required packages
import numpy as np
from numpy.linalg import svd as svd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
from collections import OrderedDict
import matplotlib.pyplot as plt

# Define plot font sizes
label_font = 18
title_font = 24
legend_font = 12

# https://pytorch.org/tutorials/intermediate/char_rnn_classification_tutorial.html
# https://blog.floydhub.com/a-beginners-guide-on-recurrent-neural-networks-with-pytorch/
# https://stackoverflow.com/questions/53819383/how-to-assign-a-new-value-to-a-pytorch-variable-without-breaking-backpropagation
```

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, net_seed=1234, I=None, W=None):
        super(RNN, self).__init__()
        # Set the sizes of layers and more.
        np.random.seed(net_seed)
        torch.manual_seed(net_seed)
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.m = nn.Parameter(torch.randn(hidden_size), requires_grad=True)
        self.n = nn.Parameter(torch.randn(hidden_size), requires_grad=True)
        self.J = Variable(torch.ger(self.m, self.n.T)/(hidden_size**0.5), requires_grad=True)

        self.I = Variable(torch.randn(hidden_size, input_size), requires_grad=False) if I is None else I
        self.W = Variable(torch.randn(output_size, hidden_size), requires_grad=False) if W is None else W

    def linear(self, W, x):
        return torch.matmul(W, x[..., None]).squeeze()

    def forward(self, C, bias, dt=0.1):
        self.J = torch.ger(self.m, self.n) / (self.hidden_size**0.5)
        n_trials = C.shape[0]
        X = [torch.zeros(n_trials, self.hidden_size).squeeze()]
        Y = []
        for i in range(C.shape[1]):
            X.append(X[-1] + self.linear(self.J, torch.tanh(X[-1])) - X[-1] * dt + torch.matmul(C[:, i], self.I.T) * np.sqrt(dt))
            Y.append(self.linear(self.W, torch.tanh(X[-1])) / self.hidden_size + 0.5)
        return torch.stack(X), torch.stack(Y)
```

```python
def train_net(network, n_trials, t_steps, cost, opti, seed_num):
    """ Performs one optimization step of batch gradient decent
    
    Parameters
        network = instance of an RNN
        n_trials = the batch size
        t_steps = the time steps for which the network runs
        cost = the loss function
        opti = the pytorch optimizer used
        seed_num = a random seed

    Returns
        loss = the computed loss for the update step
        readout = the network's response
        true_y = the desired response
    """
    
    # Set the seed
    np.random.seed(seed_num)
    torch.manual_seed(seed_num)
    
    # Define the mean c value for all batches
    c_bar = torch.rand(n_trials)
    # Define the input strengths for all time points
    cts = torch.normal(c_bar.repeat(t_steps, 1).T, c_std).unsqueeze(dim=2)
    
    # Define the optimal values
    true_y = np.heaviside(c_bar - bias, 0).squeeze()

    # Reset the gradients to zero
    opti.zero_grad()

    # Run the network for all trials
    X, readout = network(cts, bias)

    # Calculate the MSE loss
    loss = cost(readout[-1], true_y)
    # Perform backprop and calculate gradients
    loss.backward(retain_graph=True)
    # Update m1 and n1
    opti.step()

    return loss.item(), readout.detach().numpy(), true_y
   
```

```python
# Define the number of neurons in the network
N = 500

# Initialize the model
Net1 = RNN(1, N, 1)

# Double check the rank of the connectivity matrix
print('Rank of J: {}'.format(np.linalg.matrix_rank(Net1.J.detach().numpy())))
D, V = np.linalg.eigh(Net1.J.detach().numpy())

# Define the optimizer and loss function
lr = 0.001
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(Net1.parameters(), lr=lr)
```

```python
# Initialize number of batches and trials
t_steps = 200
n_batches = 1000
n_trials = 64

# Define a readout bias
bias = 0.5

# Define the stimulus strength
c_std = 0.4

# Initialize lists to store data
losses = []
readouts = []
correct = []

np.random.seed(1234)
torch.manual_seed(1234)
seeds = np.random.choice(10000, size=n_batches, replace=False)
for i in range(n_batches):
    loss, readout, labels = train_net(Net1, n_trials, t_steps, criterion, optimizer, seeds[i])
    losses.append(loss)
    readouts.append(readout)
    correct.append(np.mean(np.heaviside(readout[-1, :] - 0.5, 0) == labels.detach().numpy()))
    
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(losses)
ax.set_xlabel('Batch', fontsize=label_font)
ax.set_ylabel('Loss', fontsize=label_font)
# ax.set_title('Mean Squared Error', fontsize=title_font);
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(correct)
ax.set_xlabel('Batch', fontsize=label_font)
ax.set_ylabel('% Correct', fontsize=label_font)
# ax.set_title('Network Performance', fontsize=title_font);
```

```python
# Train 10 networks
n_nets = 10

# Initialize lists to store data
ms = np.zeros((n_nets, N))
ns = np.zeros((n_nets, N))
Is = np.zeros((n_nets, N))
ws = np.zeros((n_nets, N))

np.random.seed(1234)
torch.manual_seed(1234)
seeds = np.random.choice(10000, size=n_batches, replace=False)
for n in range(n_nets):
    # Initialize a model
    Net2 = RNN(1, N, 1)

    for i in range(n_batches):
        loss, readout, labels = train_net(Net2, n_trials, t_steps, criterion, optimizer, seeds[i])
        losses.append(loss)
        readouts.append(readout)
        correct.append(np.mean(np.heaviside(readout[-1, :] - 0.5, 0) == labels.detach().numpy()))
    
    # Collect the variables from the trained network
    ms[n, :] = Net2.m.detach().numpy()
    ns[n, :] = Net2.n.detach().numpy()
    Is[n, :] = Net2.I.detach().numpy().squeeze()
    ws[n, :] = Net2.W.detach().numpy().squeeze()
    
```

```python
# Calculate the matrices
n_params = 4
mat_comp = np.zeros((n_params, n_params))

for i in range(n_nets):
    # First row
    mat_comp[0, 0] += ms[i, :] @ ms[i, :]
    mat_comp[0, 1] += ms[i, :] @ ns[i, :]
    mat_comp[0, 2] += ms[i, :] @ Is[i, :]
    mat_comp[0, 3] += ms[i, :] @ ws[i, :]
    
    # Second row
    mat_comp[1, 0] += ns[i, :] @ ms[i, :]
    mat_comp[1, 1] += ns[i, :] @ ns[i, :]
    mat_comp[1, 2] += ns[i, :] @ Is[i, :]
    mat_comp[1, 3] += ns[i, :] @ ws[i, :]
    
    # Third row
    mat_comp[2, 0] += Is[i, :] @ ms[i, :]
    mat_comp[2, 1] += Is[i, :] @ ns[i, :]
    mat_comp[2, 2] += Is[i, :] @ Is[i, :]
    mat_comp[2, 3] += Is[i, :] @ ws[i, :]
    
    # Fourth row
    mat_comp[3, 0] += ws[i, :] @ ms[i, :]
    mat_comp[3, 1] += ws[i, :] @ ns[i, :]
    mat_comp[3, 2] += ws[i, :] @ Is[i, :]
    mat_comp[3, 3] += ws[i, :] @ ws[i, :]

# Normalize
mat_comp /= n_nets
mat_comp = np.log(abs(mat_comp))
mat_comp = mat_comp - np.min(mat_comp)
mat_comp /= np.max(mat_comp)

# Plot the matrix
tick_labels = ['', 'm', 'n', 'I', 'w']
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
mat_img = ax.matshow(mat_comp, cmap='inferno')
ax.set_xticklabels(tick_labels, fontsize=label_font)
ax.set_yticklabels(tick_labels, fontsize=label_font)
# ax.set_title('Similarity of Network Pamameters', fontsize=title_font, y=1.1)
cbar = plt.colorbar(mat_img, shrink=0.82)
cbar.ax.set_ylabel('Normalized Log of Scalar Product', fontsize=label_font);
```
