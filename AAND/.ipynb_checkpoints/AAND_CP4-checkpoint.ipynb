## 1. Spike-triggered average

### Calculate and plot the spike-triggered stimulus average (STA) for $Ï„$ in the interval $[0 , 300]$ ms:

```python
step_size = 2  # dt in [ms]
tau_period = 300  # [ms]
```

```python
def rho_to_times(rho, dt):  # rho to spike times
    data_size = len(rho)
    period = data_size*dt
    spike_times = []
    for i in range(data_size):
        if rho[i] == 1:
            spike_times = np.append(i, spike_times)
        else:
            pass
    return spike_times
```

```python
spike_times = rho_to_times(rho, step_size)  # [ms]
print(np.shape(spike_times))
```

```python
plt.eventplot(spike_times, linewidths=0.005, color='k')
plt.title('Raw Spike Train')
plt.show()
plt.plot(stim, color='k')
plt.title('Raw Stimulus')
plt.show()
```

```python
def STA(stimulus, spike_times, tau_period, step_size):
    spike_points = np.shape(spike_times)[0]  # number of points
    tau_points = int(tau_period/step_size)
    tau_range = np.linspace(0, tau_period, tau_points)  # [ms]
    C = np.zeros(tau_points)
    for n in range(tau_points):  # loop over tau
        for i in range(spike_points):  # loop over spike times
                arg = int(spike_times[i] - tau_range[n])
                C[n] += stimulus[arg]
    return C, tau_range
```

```python
spiketimes_STA = STA(stim, spike_times, tau_period, step_size)
print(np.shape(spiketimes_STA))
```

```python
plt.plot(spiketimes_STA[1], spiketimes_STA[0], color='k')
plt.title('Spike-Triggered Stimulus Average')
plt.ylabel(r'$C(\tau)$')
plt.xlabel(r'$\tau$ [ms]')
plt.xlim(tau_period, 0)
plt.grid()
plt.show()
```

### In this exercise, looping over spikes is actually much faster. Why?

Considering both processes, looping over spike times vs. the integration of rho, let's inspect the sizes of these:

```python
print(np.shape(spike_times), ', size of spike times')
print(np.shape(rho), ', size of rho')
print(np.shape(spike_times)[0]/np.shape(rho)[0], ', ratio of the two')
```

Thus, less than 10% of the computations must be done when looping over spike times here.

While I'm not taking into account the speed of these individual computations, I would assume that this is why looping over spikes is faster here.

## 2. Linear encoding kernel $D$

### Use $C(Ï„)$ to construct a linear encoding kernel $D(Ï„)$:

## <font color ='black'> $ D(\tau) = \frac{\langle r \rangle C(\tau)}{\sigma^2_s} $

```python
def linear_encoding_kernel(stimulus, spike_times, STA, step_size, trial_period):
    stim_variance = np.var(stim) * step_size  # variance of the stimulus
    spike_number = len(spike_times)  # the number of spikes
    mean_firing_rate = spike_number / trial_period  # <r> in [spikes/ms]
    encoding_kernel = mean_firing_rate * STA / stim_variance
    return encoding_kernel, mean_firing_rate
```

```python
trial_period = int(20 * 60 / 0.001)  # trial length in [ms]
encoding_kernel = linear_encoding_kernel(stim, spike_times, spiketimes_STA[0], step_size, trial_period)[0]
print(np.shape(encoding_kernel))
```

```python
plt.plot(np.linspace(0, tau_period, int(tau_period/step_size)), encoding_kernel, color='k')
plt.title('Linear Encoding Kernel')
plt.ylabel(r'$D(\tau)$')
plt.xlabel(r'$\tau$ [ms]')
plt.xlim(tau_period, 0)
plt.grid()
plt.show()
```

## 3. Predicting the neuronâ€™s response

### Convolve $D(Ï„)$ with the stimulus to compute an estimate rest of the firing rate of the H1 neuron:

## <font color='black'> $ r_{est}(t) = r_0 + \int_0^{\infty} d\tau D(\tau) s(t-\tau) $

```python
def neuron_response(kernel, stimulus):  # estimated firing rate
    return np.convolve(kernel, stimulus, mode='full')
```

```python
# want the model prediction r_est to be the same as <r>
avg_firing_rate = linear_encoding_kernel(stim, spike_times, spiketimes_STA[0], step_size, trial_period)[1]
print('<r> = ', avg_firing_rate * 1000, 'spikes per second')
```

```python
predicted_response = neuron_response(kernel=encoding_kernel, stimulus=stim)
r_0 = avg_firing_rate*1000 - np.mean(predicted_response)  # match the estimate to the actual rate
print('r_0 =', r_0, 'spikes per second')
predicted_response = r_0 + predicted_response
print('<r_est> = ', np.mean(predicted_response), 'spikes per second')
```

Rectangular window from the first sheet:

```python
def rectangular_window(tau, dt):
    w = np.zeros(len(tau))
    for i in range(len(tau)):  # Writing out the piecewise function
        if -dt/2 <= tau[i] < dt/2:
            w[i] = 1 / dt
        else:
            w[i] = 0
    return w

def rectangular_convolution(tau, data, bin_size):  # approximate firing rate
    return np.convolve(rectangular_window(tau, bin_size), data, mode='same')
```

```python
window_size = 20
window_tau = np.linspace(-trial_period/2, trial_period/2, len(rho))
function = rectangular_convolution(tau=window_tau, data=rho, bin_size=window_size)
time_bounds = [int(10000/step_size), int(12000/step_size)]  # [ms]
normalized_function = function[time_bounds[0]:time_bounds[1]] / (window_size/1000) # Normalize to get value of spikes/s
```

## 4. Extra: Real spike trains versus Poisson process

Scrapped together the following. One thing that was not clear for me was whether or not to use the spike times or rho in the autocorrelation computation. I was getting weird results with spike times, so decided to use rho. This makes more sense to me anyways, as using spike times in the function assumes that they are linked by the index number, while for rho it would just be checking the number of matching spikes at each time index.

Following code is from the provided .py file:

```python
def inh_Poisson(rates, maxRate, rates_ref, actual_train):
    '''
    start with a homogeneous Poisson process,
    and then thin to obtain inhomogeneous data
    '''
    n = np.shape(rates)[0]   # desired number of spikes
    maxRate = 150  # rate in spikes per second
    randNumbers2 = np.random.rand(n)  # n random numbers, uniform distribution
    ISIs2 = -np.log(randNumbers2)/maxRate   # ISIs in seconds
    ISIs2 = ISIs2*1000   # ISIs in milliseconds
    
    # compute spike times from ISIs
    SpikeTimes2 = np.cumsum(ISIs2)

    # thin the spike train
    randNumbers = np.random.rand(n)
    Ps = rates/maxRate
    SpikeTimes_inh = SpikeTimes2[randNumbers>Ps]
    n = np.shape(actual_train)[0]
    SpikeTimes_inh = SpikeTimes_inh[0:n]     # take just 1000 spikes

    # rates_ref = [10, 50, 100, 200, 500, 1000]  # "driving" rates in spikes per second
    ISIs_ref = np.zeros((n,len(rates_ref)))
    randNumbers3 = np.random.rand(n)  #n random numbers, uniform distribution
    t_r = 5    # absolute refractory period in ms

    for i,r in enumerate(rates_ref):
        ISIs_ref[:,i] = -np.log(randNumbers3)/r*1000 + t_r  #ISIs in milliseconds

    # compute spike times from ISIs
    SpikeTimes_ref = np.zeros((n,len(rates_ref)))
    for i,r in enumerate(rates_ref):
        SpikeTimes_ref[:,i] = np.cumsum(ISIs_ref[:,i])
    return SpikeTimes_ref
```

```python
def CV(tau):
    tau = tau[~np.isnan(tau)]  # remove nan values
    std_dev = np.std(tau)  # standard deviation
    mean = np.mean(tau)  # mean
    return std_dev / mean
```

```python
print('Coefficient of variation: %.2f, of the actual ISI' %CV(actual_ISI))
print('Coefficient of variation: %.2f, of the synthetic ISI' %CV(synthetic_ISI))
```

### (c) Compute and plot the spike-train autocorrelation function of both the actual and the synthetic spike trains over the range $[âˆ’30,30] ms$. You can use only a fraction of the trial length to speed up the computation. Explain the differences.

I recall, after the fact of writing the following, that one could use the histogram function of numpy in order to do all of this. Will be using that in the future to make things more simple.

```python
def spiketimes_to_rho(spike_times, T, dt):  # convert spike times to binary array
    rho = np.zeros(T//dt)
    steps = np.linspace(0, T, T//dt+1)
    spike_times = [int(i) for i in spike_times]
    s = set(spike_times*dt)
    for i in range(len(rho)):
        if int(steps[i]) in s:
            rho[i] = 1
        else:
            pass
    return rho
```

```python
synthetic_rho = spiketimes_to_rho(synthetic_train[0], trial_period, 2)
print(np.shape(synthetic_rho))
```

```python
def auto_correlation(data, mode):
    return np.correlate(data, data, mode)
```

```python
mode = 'same'
tenth = np.shape(rho)[0]//10  # shape of a tenth of the data set
actual_ac = auto_correlation(rho[:tenth], mode)  # actual spike train autocorrelation
synthetic_ac = auto_correlation(synthetic_rho[:tenth], mode)  # synthetic spike train autocorrelation
center = np.shape(actual_ac)[0]//2  # center of the autocorrelation
print(np.shape(actual_ac), np.shape(synthetic_ac))
```

```python
plt.plot(actual_ac, color='k', label='actual spike-train')
plt.plot(synthetic_ac, color='b', label='synthetic spike-train')
plt.ylabel('auto-correlation')
plt.xlabel('time [ms]')
plt.ylim([-50, 500])  # cut out the spike of the synthetic
plt.xlim([center-30,center+30])
plt.xticks(np.linspace(center-30, center+30, 7), np.linspace(-30, 30, 7))
plt.legend()
plt.grid()
plt.show()
```
