### Exercise H7.1: Training data

```python
def training_data(p,sigma):

    #mean vectors
    mu1 = [0,1]
    mu2 = [1,0]
    mu3 = [0,0]
    mu4 = [1,1]
    
    #multivariate normal distribution
    p0a = np.random.normal(mu1, sigma, (p//4,2))
    p0b = np.random.normal(mu2, sigma, (p//4,2))
    p1a = np.random.normal(mu3, sigma, (p//4,2))
    p1b = np.random.normal(mu4, sigma, (p//4,2))
    
    #empty data
    p0 = np.zeros((p//2,3))
    p1 = np.zeros((p//2,3))
    
    for i in range(p):
        
        #random value from the distribution size
        #choice = np.random.randint(p//4)
        choice = i
        
        if i < p//4:
        
            p0[i,0] = p0a[choice,0]
            p0[i,1] = p0a[choice,1]
            p0[i,2] = 0
            
        elif i >= p//4 and i < p//2:
            
            choice = i-30
            p1[choice,0] = p1a[choice,0]
            p1[choice,1] = p1a[choice,1]
            p1[choice,2] = 1
            
        elif i >= p//2 and i < p*3//4:
            
            choice = i-60
            p0[choice+30,0] = p0b[choice,0]
            p0[choice+30,1] = p0b[choice,1]
            p0[choice+30,2] = 0
            
        elif i >= p*3//4 and i < p:
            
            choice = i-90
            p1[choice+30,0] = p1b[choice,0]
            p1[choice+30,1] = p1b[choice,1]
            p1[choice+30,2] = 1

    return p0,p1

np.random.seed(0)
```

```python
p = 120 #size of training data
data = training_data(p,sqrt(0.1))
zero = data[0]
one = data[1]
plt.scatter(zero[:,0],zero[:,1],c='r',label='$y=0$') #points for y=0
plt.scatter(one[:,0],one[:,1],c='b',label='$y=1$') #points for y=1
plt.title('Input Samples - XOR')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xlim(-1,2)
plt.ylim(-1,2)
plt.legend()
plt.grid()
plt.show()
```

### Exercise H7.2: $k$ nearest neighbors ($k$NN)

```python
#combine all the input points into an array
x1 = np.append(zero[:,0],one[:,0])
x2 = np.append(zero[:,1],one[:,1])
yt = np.append(zero[:,2],one[:,2])
values = [x1,x2,yt]
```

```python
#for a query point and k value, return point value
def kNN(query1,query2,k,training):
    
    #query point
    query = [query1,query2]

    #determine distance in x & y directions
    distance1 = training[0] - query[0]
    distance2 = training[1] - query[1]

    #determine euclidean distance
    e_distance = np.sqrt(distance1 ** 2 + distance2 ** 2)

    #index of the minimum distance for k values
    index_min = np.argsort(e_distance)[:k]

    #determine result at the query point
    target_values = training[2][index_min[:k]]
    if np.sum(target_values)/k > 0.5:
        outcome = 1
    else: 
        outcome = 0
        
    return outcome
```

```python
def plot_kNN(points,k,training,kNN):
    
    #create points^2 number of query points
    x,y = np.linspace(-1,2,points),np.linspace(-1,2,points)
    xx,yy = np.meshgrid(x,y) #mesh the two arrays into a grid

    predicted = np.zeros((points,points))
    for i in range(points):
        for j in range(points):
            predicted[i][j] = kNN(x[i],y[j],k,training)
    
    return xx,yy,predicted
```

```python
k_values = [1,3,5]
for i in k_values:
    p_kNN = plot_kNN(90,i,values,kNN)
    scatter = plt.scatter(p_kNN[0],p_kNN[1],c=p_kNN[2],cmap='coolwarm')
    classes = ['$y=0$','$y=1$']
    plt.legend(handles=scatter.legend_elements()[0], labels=classes)
    plt.title('Nearest Neighbors $k = %i$' %i)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.xlim(-1,2)
    plt.ylim(-1,2)
    plt.grid()
    plt.show()
```

### Exercise H7.3: “Parzen window” classifier

```python
def kernel(query,training,sigma):
    arg = -(1/(2*sigma)) * ((abs(np.subtract(training,query))) ** 2)
    k = np.exp(arg)
    return k
```

```python
def parzen_window(query1,query2,sigma,training,labels):
    
    #query point
    query = [query1,query2]
    
    #one hot encoding
    labels = labels.astype(int)
    onehot = np.zeros((labels.size, labels.max()+1))
    onehot[np.arange(labels.size),labels] = 1
    
    #determine weighted vote value
    distance1 = training[0] - query[0]
    distance2 = training[1] - query[1]
    e_distance = np.sqrt(distance1 ** 2 + distance2 ** 2)
    kern = kernel(0,e_distance,sigma)
    Z = np.sum(kern)

    weighted_vote = np.argmax(np.sum((onehot.T * kern) / Z),axis=1)
    
    return weighted_vote
```

```python
# Define a parzen window classifier
def parzen_window2(var, train, test, labels):
    predictions = np.zeros(test.shape[0])

    # Perform one hot encoding
    labels = labels.astype(int)
    onehot = np.zeros((labels.size, labels.max()+1))
    onehot[np.arange(labels.size),labels] = 1

    # Find neighboor for each test sample
    for n, sample in enumerate(test):
        # Parzen window
        eucl_norm = np.sqrt(np.sum((sample-train)**2, axis=1))
        kernel = np.exp(-(eucl_norm**2) / (2*var))
        Z = np.sum(kernel)
        predictions[n] = np.argmax(np.sum((onehot.T*kernel)/Z,axis=1))

    return predictions
```

```python
x_range = np.arange(-1, 2, 0.035)
xx, yy = np.meshgrid(x_range, x_range)
x_test = np.vstack([xx.flatten(), yy.flatten()]).T

def plot_PW(sigma,classes,values):
    outcome = parzen_window2(sigma,np.transpose(values[:classes]),x_test,values[classes])
    return outcome
```

```python
sigma_values = [0.5,0.1,0.01]
for i in sigma_values:
    p_PW = plot_PW(i,2,values)
    scatter = plt.scatter(x_test[:,0],x_test[:,1],c=p_PW,cmap='coolwarm')
    classes = ['$y=0$','$y=1$']
    plt.legend(handles=scatter.legend_elements()[0], labels=classes)
    plt.title(r'Parzen Window Classifier $\sigma = %r$' %i)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.xlim(-1,2)
    plt.ylim(-1,2)
    plt.grid()
    plt.show()
```

### Rerun kNN and Parzen-window classification with a third class

```python
def third_class(p,sigma):
    
    #the mean vector
    mu = [0.5,0.5]
    
    #distribution for the third class
    third = np.random.normal(mu, sigma, (p,2))
    
    points = np.zeros((p,3)) #emptiness
    
    for i in range(p):
        points[i,0] = third[i,0]
        points[i,1] = third[i,1]
        points[i,2] = 2
    
    return points
```

```python
#for a query point and k value, return point value
def threeclass_kNN(query1,query2,k,training):
    
    #query point
    query = [query1,query2]

    #determine distance in x & y directions
    distance1 = training[0] - query[0]
    distance2 = training[1] - query[1]

    #determine euclidean distance
    e_distance = np.sqrt(distance1 ** 2 + distance2 ** 2)

    #index of the minimum distance for k values
    index_min = np.argsort(e_distance)[:k]

    #determine result at the query point
    target_values = training[2][index_min[:k]]
    
    outcome = max(target_values)
        
    return outcome
```

```python
def basis_function(query,sigma,K,cluster_range):
    
    #use kmeans to determine representatives
    for i, k in enumerate(cluster_range):
        kmeans = KMeans(n_clusters=k, random_state=0).fit(data[0])
        centers = kmeans.cluster_centers_
    
    t = [centers[:, 0], centers[:, 1]]
    
    q = []
    [q.append(np.transpose(query)) for i in range(len(t[0]))]
    q = np.reshape(q,(2,K))
    
    phi = kernel(q,t,sigma)
    
    return phi,centers
```

### Determine the weight vector and plot

```python
def weight_vector(query,yT,sigma,K,test,train):
    
    cluster_range = [2,3,4]
    
    phi = basis_function(query,sigma,K,cluster_range)[0]
    centers = basis_function(query,sigma,K,cluster_range)[1]
    
    #a = np.dot(phi,np.linalg.inv(np.dot(phi,phi.T)))
    #print(np.shape(a))
    #w = np.dot(yT.T,a)
    
    
    phi_train = np.ones(train.shape[0])
    phi_test = np.ones(test.shape[0])
    for i, ti in enumerate(centers):
        eucl_norm = np.sqrt(np.sum((train-ti)**2, axis=1))
        phi = np.exp(-((eucl_norm)**2) / (2*var))
        phi_train = np.vstack([phi_train, phi])

        eucl_norm = np.sqrt(np.sum((test-ti)**2, axis=1))
        phi = np.exp(-((eucl_norm)**2) / (2*sigma))
        phi_test = np.vstack([phi_test, phi])
    
    #calculate the weight
    weight = np.matmul(np.linalg.inv(np.matmul(phi, phi.T)), np.matmul(phi, yT))
    
    return weight,phi
```
