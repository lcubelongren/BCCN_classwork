## Exercise 8.1: K-means Clustering – batch version

### A: Initialization –

```python
"""
1. Set the initial position of each prototype w_q randomly around the mean of the entire dateset.
2. Set the maximum number of iterations: t_max = 5
"""
def batch_initial(data, M, t_max):
    mean = np.mean(data)  # mean of entire dataset
    w_q = np.zeros((t_max+1, 2, M))
    rand_mag = np.max(data)/10  # magnitude of randomness
    for i in range(M):  # random initial prototypes
        w_q[0][0][i] = mean + np.random.uniform(-abs(rand_mag), abs(rand_mag))
        w_q[0][1][i] = mean + np.random.uniform(-abs(rand_mag), abs(rand_mag))
    return w_q, t_max
```

### B: Optimization –

```python
"""
Implement the K-means update. Each iteration should contain the following two steps
1. Assign all datapoints to their closest prototype.
2. Re-compute the location of the prototypes due to the new assignments.
"""
def batch_optim(data, M, t_max):
    w_q, t_max = batch_initial(data, M, t_max)  # initialize prototypes and iteration max
    data_num = np.shape(data)[1]  # number of data points
    m_q = np.zeros((t_max+1, data_num, M))  # assignment of data to prototype
    for t in range(t_max):  # loop over iterations 
        for i in range(data_num):  # loop over data
            distance = np.linalg.norm(data[:,i] - np.transpose(w_q[t]), axis=1)  # euclidean distance
            closest_idx = np.argmin(distance)  # index of closest prototype
            m_q[t+1][i][closest_idx] = 1  # assign closest cluster to each data points
        # now minimize the error function to assign new w_q
        # set it to the center of mass of found m_q
        for j in range(M):  # update prototypes
            w_q[t+1][0][j] = np.sum(m_q[t][:,j] * data[0]) / (np.sum(m_q[t][:,j]) + 0.1)
            w_q[t+1][1][j] = np.sum(m_q[t][:,j] * data[1]) / (np.sum(m_q[t][:,j]) + 0.1)
            # added the 0.1 to the denominator to avoid overflow erros
    return w_q, m_q  # return prototypes for all iterations (trial, x1/x2, batch)
```

### C: Visualization –

### (a) Visualize data points and prototypes for each iteration in a sequence of scatter plots.

```python
M = np.arange(2, 10, 1)  # cluster numbers (did M=9 also for nicer plotting)
t_max = 5  # iteration number

fig, axs = plt.subplots(2, len(M)//2, figsize=(25,15), sharex=True, sharey=True)
axs = axs.ravel()
cmap = plt.cm.viridis
for i in range(len(M)):  # plot clusters
    w_q, m_q = batch_optim(data, M=M[i], t_max=t_max)  # calculate all iterations here
    axs[i].scatter(data[0], data[1], color='k', alpha=0.6)  # plot raw data
    axs[i].set_xlabel('$x_1$'), axs[i].set_ylabel('$x_2$')
    
    color = iter(cmap(np.linspace(0, 1, t_max)))
    for j in range(t_max):  # plot iterations
        c = next(color)
        axs[i].scatter(w_q[j][0][:], w_q[j][1][:], color=c, s=2**9)
        axs[i].set_title('M = {0}'.format(M[i]))
        axs[i].grid()
plt.tight_layout()
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=plt.cm.colors.Normalize(), cmap=cmap), ax=axs, 
                    boundaries=np.linspace(-0.5, t_max+0.5, t_max+2), ticks=np.arange(0, t_max+1, 1),
                    orientation='horizontal', label='iteration #', shrink=0.75, aspect=40, pad=0.05)
plt.show()
```

### (b) For two different initializations, plot the error function $E$ vs. the iteration $t$:

# <font color='k'> $ E_{[\{m_q^{(\alpha)}\}, \{\bar{w}_q \}]} = \frac{1}{p} \sum_{\alpha=1}^p \sum_{q=1}^M m_q^{(\alpha)} || \bar{x}^{\alpha} - \bar{w}_q || ^2_2$

```python
def error_function(data, M, t_max):
    w_q, m_q = batch_optim(data, M=M, t_max=t_max)  # calculate all iterations here
    error = np.zeros(t_max+1)
    for t in range(t_max+1):  # loop over iterations
        for i in range(np.shape(data)[1]):  # loop over data points (alpha)
            for j in range(M):  # loop over batches (M)
                error[t] += m_q[t][i][j] * (abs(data[0][i] - w_q[t][0][j]) ** 2 + abs(data[1][i] - w_q[t][1][j]) ** 2)
    return error * (1/np.shape(data)[1])
```

```python
plt.plot(error_function(data, M=3, t_max=5), label='$M=3$', color='g', marker='o')
plt.plot(error_function(data, M=7, t_max=5), label='$M=7$', color='b', marker='o')
plt.title('Error Function')
plt.xlabel('iteration #'), plt.ylabel('error')
plt.legend()
plt.grid()
plt.show()
```

### (c) Create a plot (Voronoi-Tesselation) to show how the resulting solution would potentially assign new data points 
### (i.e. show the decision boundaries that separate the clusters).

```python
w_q, m_q = batch_optim(data, M=5, t_max=t_max)  # calculate all iterations here
for i in range(t_max+1):
    #plt.scatter(data[0], data[1], color='k', alpha=0.6)  # plot raw data
    vor = Voronoi(np.transpose([w_q[i][0][:], w_q[i][1][:]]))
    fig = voronoi_plot_2d(vor)
    plt.show()
```
