### a) Simulate the Hodgkin-Huxley model by numerically integrating the equations

First, let us define all of the functions given to us:

### The Hodgkin-Huxley Model

```python
def HHM(I_na,I_k,I_leak,I_e,C_m):
    dvdt = (1/C_m) * (-(I_na + I_k + I_leak) + I_e)
    return dvdt
```

### Currents $I_{K}$ & $I_{Na}$

```python
def current_k(V,n,g_k,E_k):
    current_k = g_k * (n**4) * (V - E_k)
    return current_k
```

```python
def current_na(V,m,h,g_na,E_na):
    current_na = g_na * (m**3) * h * (V - E_na)
    return current_na
```

### First-order Kinetics $\frac{dx}{dt}$

```python
def kinetics(x,alpha,beta):
    dxdt = alpha * (1 - x) - beta * x
    return dxdt
```

### Transition Rates $\alpha$ & $\beta$

```python
def alpha_n(V):
    alpha = 0.01 * (V + 55) / (1 - np.exp(-0.1 * (V + 55)))
    return alpha
```

```python
def alpha_m(V):
    alpha = 0.1 * (V + 40) / (1 - exp(-0.1 * (V + 40)))
    return alpha
```

```python
def alpha_h(V):
    alpha = 0.07 * exp(-0.05 * (V + 65))
    return alpha
```

```python
def beta_n(V):
    beta = 0.125 * np.exp(-0.0125 * (V+65))
    return beta
```

```python
def beta_m(V):
    beta = 4 * exp(-0.0556 * (V + 65))
    return beta
```

```python
def beta_h(V):
    beta = 1 / (1 + exp(-0.1 * (V + 35)))
    return beta
```

### Passive Leakage Current $I_{leak}$

```python
def current_leak(V,g_leak,E_leak):
    current = g_leak * (V - E_leak)
    return current
```

Now we shall integrate the ODEs. This time, let's try our hand at the $odeint$ function from the scipy package.

```python
#the 4 ODEs to be integrated
def HHM_combined(initial, t, C_m, I_e, E_na, E_k, E_leak, g_na, g_k, g_leak, I_following): 
    
    V,n,m,h = initial #initial values
    
    #calculate potassium current
    kinetics_n = kinetics(n,alpha_n(V),beta_n(V))
    I_k = current_k(V,n,g_k,E_k)
    
    #calculate sodium current
    kinetics_m = kinetics(m,alpha_m(V),beta_m(V))
    kinetics_h = kinetics(h,alpha_h(V),beta_h(V))
    I_na = current_na(V,m,h,g_na,E_na)
    
    #calculate leak current
    I_leak = current_leak(V,g_leak,E_leak)
    
    #calculate voltage 
    if t <= duration:
        voltage = HHM(I_na,I_k,I_leak,I_following,C_m)
    else:
        voltage = HHM(I_na,I_k,I_leak,I_e,C_m)
    
    return [voltage,kinetics_n,kinetics_m,kinetics_h]
```

```python
#initial conditions
V0 = -65 #(mV)
m0 = 0.0529
h0 = 0.5961
n0 = 0.3177
#parameters
C_m = 1 #(pF)
I_e = 0 #(A)
E_na = 50 #(mV)
E_k = -77 #(mV)
E_leak = -54.387 #(mV)
g_na = 120 #(nS)
g_k = 36 #(nS)
g_leak = 0.3 #(nS)

a = 
g = a / (2 * rL * L * L)

#time steps
t = np.linspace(0,50,1e5)
#for later
I_following = I_e
duration = 5
```

```python
#plot the solution to the HHM
I_e = 0 #(A)
initial = [V0,n0,m0,h0]
parameters = (C_m, I_e, E_na, E_k, E_leak, g_na, g_k, g_leak, I_following)
solution = odeint(HHM_combined, initial, t, parameters)
plt.plot(t,solution[:,0])
plt.title('Membrane Voltage with $I_e = 0$')
plt.ylabel('voltage [mV]')
plt.xlabel('time [t]')
plt.grid()

#fix offset
ax = plt.gca()
ax.ticklabel_format(useOffset=False)
plt.show()
```

At different phases of the action potential, the different currents are influenced by the dynamics of the gating variables. For the potassium current, which is a function of the gating variable $n$, as the gating variable increases so does the current. For the sodium current, which is a function of the gating variables $m$ and $h$, as $m$ increases and $h$ decreases the current increases.

Our intuition of this relationship can be reinforced by Dayan and Abbott, where on p.172 they state "[t]he activation and inactivation variables $m$ and $h$ are distinguished by having opposite voltage dependences. Depolarization causes $m$ to increase and $h$ to decrease, and hyperpolarization decreases $m$ while increasing $h$.

For the plots of current vs voltage and gating variable $n$ vs voltage, these illustrations are of limit cycles which is defined as a close trajectory in phase space in which another trajectory spirals into it as time approaches infitity or negative infitity. In other words, in our case as time goes to infinity the relationship between gating variable and voltage (hence current and voltage) approaches a constant repeated pattern after a short time.

### c) Plot the firing rate of the model

```python
#let's create the following firing rate function
def firing_rate(I_e,E_m,points):
    firing_rate = np.zeros(I_range.shape[0])
    rheobase = []
    for i in range(points):
        t = np.linspace(0,1000,1e3) #time range for the HHM
        parameters = (C_m, I_e[i], E_na, E_k, E_leak, g_na, g_k, g_leak, I_following)
        voltage = odeint(HHM_combined, initial, t, parameters)[:,0]
        #calculate where the HHM peaks above E_m
        firing_rate[i] = len(find_peaks(voltage, height=E_m)[0])
        #determine the value of the rheobase
        if firing_rate[i] > 40:
            rheobase.append(I_e[i])
    return firing_rate,rheobase
```

```python
I_range = np.linspace(0,20,200)
firing_rate = firing_rate(I_range,-40,200)
```

```python
print('approximate rheobase value: I = %r pA' %firing_rate[1][0])
plt.plot(I_range,firing_rate[0])
plt.axvline(x = firing_rate[1][0],c='k',linestyle='--',label='rheobase ~ 6.3 pA')
plt.title('Firing Rate of the Model vs Current')
plt.ylabel('firing rate [Hz]')
plt.xlabel('current [pA]')
plt.legend()
plt.grid()
plt.show()
```

We can see from the above plot that the apporoximate rheobase value is 6.3 pA. This point is where the firing rate jumps discontinuously from zero to a nonzero value of a little under 60 Hz. The discontinuous jump is an example of a type 2 firing rate. This is contrasted to the previous firing rate of the leaky integrate-and-fire model of which has a continuous increase from zero to a non-zero firing rate. That was an example of a type 1 firing rate.

### d) Apply a negative current pulse

```python
t = np.linspace(0,50,1e3)
#for all points in t, separate current values I_e
#into the two different values -5pA and 0
I_e = 0 #(pA) current for duration
I_following = -5 #current after duration
duration = 5 #(ms) 
parameters = (C_m, I_e, E_na, E_k, E_leak, g_na, g_k, g_leak, I_following)
solution = odeint(HHM_combined, initial, t, parameters)
```

```python
_, ax = plt.subplots(3,1,figsize=(15,5))
ax[0].plot(t,solution[:,0])
ax[0].ticklabel_format(useOffset=False)
ax[0].set_title('Membrane Voltage vs Time')
ax[0].set_ylabel('voltage [mV]')
ax[0].set_xlabel('time [ms]')
ax[0].grid()

ax[1].plot(t,solution[:,1],label='$n$')
ax[1].plot(t,solution[:,2],label='$m$')
ax[1].plot(t,solution[:,3],label='$h$')
ax[1].set_title('Gating Variables vs Time')
ax[1].set_ylabel('$n$,$m$,$h$')
ax[1].set_xlabel('time [ms]')
ax[1].legend()
ax[1].grid()

c_k = current_k(solution[:,0],solution[:,1],g_k,E_k)
c_na = current_na(solution[:,0],solution[:,2],solution[:,3],g_na,E_na)
ax[2].plot(t,c_k,label='$I_K$')
ax[2].plot(t,-c_na,label='$-I_{Na}$')
ax[2].set_title('Current vs Time')
ax[2].set_ylabel('current [pA]')
ax[2].set_xlabel('time [ms]')
ax[2].legend()
ax[2].grid()

plt.tight_layout()
plt.show()
```

Above we see plots of membrane voltage, gating varaibles, and current as a function of time in reaction to an applied negative current pulse for a time period of 5ms. We can see that @ $t=5ms$, the instantanious jump from -5pA to 0pA induces a significant change in the the value for the gating variables (of which are a function of voltage). This has an increase in magnitude of each current (of which is a function of the gating variables) until a point shortly after $t=10ms$. At this point, the modeled neuron hits a threshold in the membrane voltage and sees an action potential spike induced. After the characteristic spike, the membrane voltage undershoots, and then returns to its resting state.

Compared to the leaky integrate-and-fire neuron, where when this threshold value was reached we 'artificially' induced a spike and returned the model neuron to the resting state manually, here for the Hodgkin-Huxley model instead the action potential spike and return to rest are built into the model via the 4 coupled ODEs.
