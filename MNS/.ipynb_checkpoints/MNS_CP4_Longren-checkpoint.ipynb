# Ordinary Differential Equations - Luke Longren

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from math import *

%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

## 1. Analytical solutions to ODEs

See attached PDF for the derivation.

```python
#problem (a)
def xt_a_analytical(t): 
    return exp(-t)

#problem (b)
def xt_b_analytical(t):
    return sqrt(2*t + 1)

#problem (c)
def xt_c_analytical(t):
    return 1 - exp(-t)

#problem (d)
def xt_d_analytical(t):
    return 1 / (1 + exp(-t))
```

## 2. Numerical solutions to ODEs

### a) Create the Euler function

```python
#as a function of the function, initial x, max t, and delta t

def euler(fx,x0,t_max,t_del): 
    
    x = np.zeros(t_max) #emptiness
    x[0] = x0 #initial condition
    
    for i in range(t_max-1):
        
        x[i+1] = x[i] + (fx(x)[i] * t_del)

    return x
```

### b) Solve the problems using Euler's

```python
t_max = 50 #the stop time
t_del = 0.25 #the integration step

#problem (a)

def fx_a(x):
    return -x

x0_a = 1
xt_a = euler(fx_a,x0_a,t_max,t_del) #the numerical solution

t_a = np.linspace(0,t_max*t_del,t_max) #the analytical solution
xt_a_analytical = [xt_a_analytical(t_a[i]) for i in range(t_max)]

plt.plot(t_a,xt_a,label="euler's")
plt.plot(t_a,xt_a_analytical,label="analytical")
plt.title('x* = -x')
plt.ylabel('x')
plt.xlabel('t')
plt.legend()
plt.grid()
plt.show()
```

```python
# problem (b)

def fx_b(x):
    return 1 / (x + 1e-5) #avoid divide by zero with something small

x0_b = 1
xt_b = euler(fx_b,x0_b,t_max,t_del) #the numerical solution

t_b = np.linspace(0,t_max*t_del,t_max) #the analytical solution
xt_b_analytical = [xt_b_analytical(t_b[i]) for i in range(t_max)]

plt.plot(t_b,xt_b,label="euler's")
plt.plot(t_b,xt_b_analytical,label="analytical")
plt.title('x* = 1/x')
plt.ylabel('x')
plt.xlabel('t')
plt.legend()
plt.grid()
plt.show()
```

```python
#problem (c) 

def fx_c(x):
    return 1 - x

x0_c = 0
xt_c = euler(fx_c,x0_c,t_max,t_del) #the numerical solution

t_c = np.linspace(0,t_max*t_del,t_max) #the analytical solution
xt_c_analytical = [xt_c_analytical(t_c[i]) for i in range(t_max)]

plt.plot(t_c,xt_c,label="euler's")
plt.plot(t_c,xt_c_analytical,label="analytical")
plt.title('x* = 1 - x')
plt.ylabel('x')
plt.xlabel('t')
plt.legend()
plt.grid()
plt.show()
```

```python
#problem (d)

def fx_d(x):
    return x * (1 - x)

x0_d = 0.5
xt_d = euler(fx_d,x0_d,t_max,t_del) #the numerical solution

t_d = np.linspace(0,t_max*t_del,t_max) #the analytical solution
xt_d_analytical = [xt_d_analytical(t_d[i]) for i in range(t_max)]

plt.plot(t_d,xt_d,label="euler's")
plt.plot(t_d,xt_d_analytical,label="analytical")
plt.title('x* = x(1 - x)')
plt.ylabel('x')
plt.xlabel('t')
plt.legend()
plt.grid()
plt.show()
```

For our reasonable step size, we are obtaining a very good approximation for these ODEs using Euler's numerical method of calculating the derivative. For each of the ODEs that we have examined here, one can see that the numerical method is over-approximating the functions, returning a result that has a larger curve than the analytical solution. 

### c) Comparing with Runge-Kutta, order 5(4)

```python
#let's take a look at the highest order equation, problem (d)

def rungekutta(fx,x0,t_max,t_del):
    
    x = np.zeros(t_max) #emptiness
    x[0] = x0 #initial condition
    
    for i in range(t_max-1): #going through the process of calculating RK4
        
        #finding each of the increments k
        k1 = t_del * fx(x)[i]
        k2 = t_del * fx(x + (k1/2))[i]
        k3 = t_del * fx(x + (k2/2))[i]
        k4 = t_del * fx(x + k3)[i]
        
        x[i+1] = x[i] + (1/6) * (k1 + 2*k2 + 2*k3 + k4) #bringing it together

    return x


N = 4 #plotting stuff
fig, axs = plt.subplots(1, N, figsize=(14, 3), sharey=True)
for i in range(N):
    
    t_max = 50 #the stop time
    t_del = 0.1*(i+1) #the integration step
    tit = "{:.5g}".format(0.1*(i+1)) #string for the plotting title

    xt_d_RK = rungekutta(fx_d,x0_d,t_max,t_del) #the runge-kutta numerical method
    xt_d = euler(fx_d,x0_d,t_max,t_del) #compared to euler's with the same steps
    
    #more plotting stuff
    
    axs[i].plot(t_d,xt_d_RK,label="RK4")
    axs[i].plot(t_d,xt_d,label="euler's")
    axs[i].plot(t_d,xt_d_analytical,label="analytical")
    axs[i].set_title('step-size = %s' %tit)
    axs[i].set_ylabel('x')
    axs[i].set_xlabel('t')
    axs[i].legend()
    axs[i].grid()
    fig.suptitle('x* = x(1 - x)',fontsize=14,va='baseline')

plt.show()
```

One can see that as the step size is increased past about 0.2, the numerical solutions go from undershooting to overshooting the analytical solution for this ODE. Also, for for small step sizes it looks like Euler's is a better method than our RK4. And for the larger step sizes, RK4 better approximates the analytical solution of our ODE.

### 3. Passive membrane

### a) Implement the equation describing a passive membrane subject to a current injection

```python
#passive membrane with a current injection function

R = 10**7 #membrane resistance (ohmes)
I0 = 10**-6 #injected current (mA)
tau = 10 #membrane constant (msec)
E = -80 #membrane voltage at resting state (mV)

def membrane_single(V): #derivative of voltage w/ respect to time
    
    voltage_der = (-V / tau) + (E / tau) + ((R * I0) / tau)
    
    return voltage_der
```

```python
#euler method to solve this ODE

t_max = 10**4 #the stop time
t_del = 0.01 #the integration step

voltage = euler(membrane_single,E,t_max,t_del)

t = np.linspace(0,t_max*t_del,t_max)
plt.plot(t,voltage)
plt.title('passive membrane subject to current injection @ t=0')
plt.ylabel('voltage (mV)')
plt.xlabel('time (ms)')
plt.grid()
plt.show()
```

### b) Consider a sinusoidal time-dependent current

```python
#as a function of the function, initial x, max t, and delta t

def euler_new(fx,x0,t_max,t_del): #new euler that can take an ODE dependent on time
    
    t = np.linspace(0,t_max*t_del,t_max)
    x = np.zeros(t_max) #emptiness
    x[0] = x0 #initial condition
    
    for i in range(t_max-1):
        
        x[i+1] = x[i] + (fx(x,t[i])[i] * t_del)

    return x
```

```python
#with a sinusoidal current

def membrane_sine(V,t): 

    I = I0 * sin(2*pi*nu*t)
    voltage_der = (-V / tau) + (E / tau) + ((R * I) / tau)
    
    return voltage_der
```

```python
#function for the rescaled input

def rescaled_input():

    t = np.linspace(0,t_max*t_del,t_max)
    rs = np.zeros(len(t))

    for i in range(100):
        
        t0 = t[i] 
        rs[i] = E + R * (I0 * sin(2*pi*nu*t0)) 
    
    return rs
```

```python
R = 10**7 #membrane resistance (ohmes)
I0 = 10**-9 #injected current (ampes)
tau = 0.01 #membrane constant (seconds)
E = -0.08 #membrane voltage at resting state (volts)
```

```python
#euler method to solve this ODE

t_max = 2000 #the stop time
t_del = 0.005 #the integration step

nu = 1 #frequency (Hz)

t = np.linspace(0,t_max*t_del,t_max)
plt.plot(t,euler_new(membrane_sine,E,t_max,t_del),label='voltage') #the sinusoid
plt.plot(t,rescaled_input(),label='rescaled input') #the rescaled input
plt.title('sinusoidal current, nu = %i' % nu)
plt.ylabel('voltage (V)')
plt.xlabel('time (s)')
plt.legend()
plt.grid()
plt.show()
```

### c) Plot the amplitude and phase of the resulting voltage oscillation versus frequency

```python
#the analytical functions defined

def amplitude_analytical(nu):
    
    amplitude = I0 * R * sqrt(1/(1 + (2*pi*nu*tau)**2))
    
    return amplitude

def phase_analytical(nu):
    
    phase = np.arctan(2*pi*nu*tau)
    
    return phase
```

```python
#now let's go back and do it numerically first

#create starting arrays for values, size of the number of frequency values
nu0 = np.zeros(100) #this will be for frequency
voltage0 = [[] for _ in range(100)] #this will be for voltages, peak not found
phase0 = np.zeros(100) #this will be for the index of the peak

for i in range(100):
    
    nu0[i] = 1 + i #[1Hz,100Hz]
    nu = nu0[i]

    voltage0[i] = euler_new(membrane_sine,E,t_max,t_del) #voltage array for single nu
    
    phase0[i] = np.argmax(voltage0[i]) #phase index of the resulting oscillation   
    
voltage = np.zeros(100) #this will be for voltages, peak found

for i in range(100):

    phase = int(phase0[i]) 
    voltage[i] = voltage0[i][phase] #this finds the voltage at maximums
    
plt.plot(nu0,voltage) #plotting amplitude as function of frequency
plt.title('amplitude of the voltage oscillation versus frequency')
plt.ylabel('amplitude of voltage (V)')
plt.xlabel('frequency (Hz)')
plt.grid()
plt.show()
```

```python
#now to do phase versus frequency, which is quite easy with what we already have
#from the previous cell, 'phase0' is the index of the first peak 
plt.plot(nu0,phase0) #plotting phase as function of frequency
plt.title('phase of the voltage oscillation versus frequency')
plt.ylabel('phase of voltage (rad)')
plt.xlabel('frequency (Hz)')
plt.grid()
plt.show()
```

```python
#back to the analytical solution
a_analytical = np.zeros(100)
for i in range(100):
    a_analytical[i] = amplitude_analytical(nu0[i]) 
    
plt.plot(nu0,a_analytical) #plotting analytical amplitude as function of frequency
plt.title('analytical amplitude versus frequency')
plt.ylabel('amplitude of voltage (V)')
plt.xlabel('frequency (Hz)')
plt.grid()
plt.show()
```

```python
p_analytical = np.zeros(100)
for i in range(100):
    p_analytical[i] = phase_analytical(nu0[i]) 
    
plt.plot(nu0,p_analytical) #plotting analytical phase as function of frequency
plt.title('analytical phase versus frequency')
plt.ylabel('phase of voltage (rad)')
plt.xlabel('frequency (Hz)')
plt.grid()
plt.show()
```

```python
#let's now compare the two values on the same graph
plt.plot(nu0,a_analytical,label='analytical') #analytical amplitude
plt.plot(nu0,voltage,label='numerical') #numerical amplitude
plt.title('amplitude versus frequency')
plt.ylabel('amplitude of voltage (V)')
plt.xlabel('frequency (Hz)')
plt.legend()
plt.grid()
plt.show()

plt.plot(nu0,p_analytical,label='analytical') #analytical phase
plt.plot(nu0,phase0,label='numerical') #numerical phase
plt.title('phase versus frequency')
plt.ylabel('phase of voltage (rad)')
plt.xlabel('frequency (Hz)')
plt.legend()
plt.grid()
plt.show()
```

Plotting the resulting amplitude and frequencies we can see that for mine the amplitude numerical and analytical plots match, besides a phase shift of the numerical about 80 volts below. For the phase plots, I find that the numerical is way overestimating the phase for lower frequencies.
