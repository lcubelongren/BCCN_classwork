```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
## H2.2 a)

c = 100 #number of data points across bounds
N = 10 #number of hidden units
x = np.linspace(-2,2,c) #define the bounds

def mlp(): #create the independent MLP
    
    y = [] 
    x = np.linspace(-2,2,c) 
    N = 10 
    
    for i in range(1,N+1): 
        
        w21 = np.random.normal(0,1) #normal dist.
        w10 = np.random.normal(0,2) #normal dist.
        b = np.random.uniform(-2,2) #uniform dist.
        
        y0 = w21 * np.tanh(w10 * x - b) #calculate an y(x) for a single i
        y.append(y0)
    
    y = sum(y[0:N]) #summation of y_i
    return y

#plt.plot(mlp(y))
```

```python
## H2.2 b)

U = 50 #number of MLPs

def joint(): #run multiple MLPs
    
    for _ in range(U):
        
        mlp()
        plt.plot(x, mlp())
        
joint()

plt.title('input-output functions for 50 MLPs where w10_i1 ∼ (0, 2)')
plt.xlabel('x')
plt.ylabel('Response y(x)')
plt.show()
```

```python
## H2.2 c)

def mlp2(): #create the independent MLP
    
    y2 = [] 
    x = np.linspace(-2,2,c) 
    N = 10 
    
    for i in range(1,N+1): 
        
        w21 = np.random.normal(0,1) #normal dist.
        w10 = np.random.normal(0,0.5) #normal dist.
        b = np.random.uniform(-2,2) #uniform dist.
        
        y0 = w21 * np.tanh(w10 * x - b) #calculate an y(x) for a single i
        y2.append(y0)
    
    y2 = sum(y2[0:N]) #summation of y_i
    return y2


U = 50 #number of MLPs

def joint2(): #run multiple MLPs
    
    for _ in range(U):
        
        mlp2()
        plt.plot(x, mlp2())
        
joint2()

plt.title('input-output functions for 50 MLPs where w10_i1 ∼ (0, 0.5)')
plt.xlabel('x')
plt.ylabel('Response y(x)')
plt.show()
```

The first plot in part a) had a weight w10_i1 normally distributed on ∼ N (0, 2). Now for part c), we make this weight w10_i1 normally distributed ∼ N (0, 0.5).

The result is that the second graph (part c)) has a much more flat plot when compared against the first graph (part a)) which had a more defined curve that resembled the activation function tanh(x).

```python
## H2.2 d)

U = 50 #number of MLPs
g = -x #function g(x)

e = np.array([np.array((mlp() - g) ** 2) for _ in range(U)]) #finding the mean squared error from a)
e = e.mean(axis=0)

plt.plot(x, e)
    
plt.show()    

e2 = np.array([np.array((mlp2() - g) ** 2) for _ in range(U)]) #finding the mean squared error from c)
e2 = e2.mean(axis=0)

plt.plot(x, e2)

plt.show()
```

Comparing the two procedures, the MLP that approximates g(x) the best would be the MLP from part a) as it has much more linear values than the values from c), where g(x) is a linear function
