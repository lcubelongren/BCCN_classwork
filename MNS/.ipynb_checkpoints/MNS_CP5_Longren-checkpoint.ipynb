## 1. Synaptic current

```python
#the given constants
tau_syn = 0.01 #(seconds)
tau_m = 0.01 #(seconds)
R_m = 1e7 #(ohmes)
g_dot = 30e-9 #(siemens)
E_m = -0.08 #(volts)
```

### a) Single pre-synaptic spike

```python
#the euler function we created last problem set
def euler(fx,x0,t_max,t_del): 
    
    x = np.zeros(t_max) #emptiness
    x[0] = x0 #initial condition

    for i in range(t_max-1):
        
        x[i+1] = x[i] + (fx(x,i)[i] * t_del)
        
    return x
```

```python
t_max = 200 #the stop time
t_del = 0.0005 #the integration step
t_spike = 0.25 * t_max #where the spike is
```

Here below we define all of the functions that we will then later solve numerically.

```python
def variable_x(x,count):
    t = count
    if t == t_spike:
        #dirac_delta = g_dot
        dirac_delta = 1 / t_del #integration must equal 1
    else:
        dirac_delta = 0
    var_x = (-x + (tau_syn*g_dot*dirac_delta)) / tau_syn

    return var_x
```

```python
def synaptic_conductance(g_syn,count):
    g_syn = (var_x - g_syn) / tau_syn
    return g_syn
```

```python
def membrane_potential(V_t,count):
    V_t = (-V_t + E_m - (R_m * (syn_conductance * (V_t - E_syn)))) / tau_m
    return V_t
```

```python
def synaptic_current():
    I_syn = syn_conductance * (mem_potential - E_syn)
    return I_syn
```

```python
def membrane_current():
    I_m = (mem_potential - E_m) / R_m
    return I_m
```

```python
#voltage values for each of the two cases
#inhibitory & excitatory
E_inh = -0.1 #inhibitory (volts)
E_exc = 0 #excitatory (volts)
```

### b) Shunting inhibition

```python
#given constants for conductance
g_exc = 50e-9 #(siemens)
g_inh = 200e-9 #(siemens)
```

```python
def total_potential(V_t,count): #potential with excitatory and inhibitory inputs also
    I_syn = syn_conductance * (V_t - E_syn)
    I_exc = g_exc * (V_t - E_exc)
    I_inh = g_inh * (V_t - E_inh)
    total_current = I_syn + I_exc + I_inh
    V_t = (-V_t + E_m - (R_m * total_current)) / tau_m
    return V_t
```

```python
tot_potential = euler(total_potential,E_m,t_max,t_del) #V(0) = E_m

#plotting the two membrane potentials-----

plt.figure(figsize=[12,8])
plt.plot(t,tot_potential,label='total membrane potential',linestyle='-')
plt.plot(t,mem_potential,label='previous membrane potential',linestyle='--')
plt.title('membrane potentials')
plt.ylabel('membrane potential (V)')
plt.xlabel('time (s)')
plt.legend()
plt.grid()
plt.show()
```

Here we have shown that the resting potential does not change due to the tonic synaptic inputs (both come back to a resting value of -80mV). Shunting inhibition, of which we are doing here, is also known as divisive inhibition. This reduction in potential is done by division instead of subtraction as biologically currents are generated at adjacent excitatory synapses. Thus, adding the extra terms for more inputed current inhibit the neuron from reaching as high a potential as our previous model.

## 2. Integrate-and-fire neuron

### a) Modified Euler method

```python
#the new modified Euler method
def modified_euler(fx,x0,t_max,t_del): 
    
    x = np.zeros(t_max) #emptiness
    x[0] = x0 #initial condition
    
    sp_count = 0 #number of spikes
    time = 0 #current time point
    
    x_thr = -0.054 #threshold (V)
    x_res = E_m #reset value
    #x_res = V_res #reset value

    for i in range(t_max-1):
        
        time += 1
        
        #if the value has reached threshold, reset
        #otherwise continue as usual 
        if x[i] >= x_thr:
            
            x[i+1] = x_res
            sp_count += 1
            
        else:
            
            x[i+1] = x[i] + (fx(x,i)[i] * t_del)
        
    return x,sp_count,time
```

### b) Simulated constant input current neuron

```python
def integrate_potential(V_t,count): #integrate-and-fire potential
    V_t = (-V_t + E_m + (R_m * I_e)) / tau_m #with I_syn(t) = 0
    return V_t
```

```python
I_e = 3e-9 #(amps)
R_m = 1e7
#other random initial conditions
E_m = -0.08
V_res = E_m
```

```python
t_max = 200 #the stop time
t_del = 0.0005 #the integration step

int_potential = modified_euler(integrate_potential,E_m,t_max,t_del)

#plotting the membrane potential for 100ms-----

t = np.linspace(0,t_max*t_del,t_max) 
plt.figure(figsize=[12,8])
plt.plot(t,int_potential[0])
plt.title('membrane potential')
plt.ylabel('membrane potential (V)')
plt.xlabel('time (s)')
plt.grid()
plt.show()
```

```python
#estimated firing rate in spikes per second
def firing_rate():
    sp_count = int_potential[1]
    time_window = t_max * t_del
    return sp_count / time_window
print('firing rate for 100ms:',firing_rate())
```

```python
#inter-spike interval (ISI) function
def ISI():
    int_potential = modified_euler(integrate_potential,E_m,t_max,t_del)[0]
    isi = 0
    for i in range(1,t_max):
        if int_potential[i] != E_m: 
            isi += 1 #if potential is not a spike, add one
        else:
            break #otherwise we have counted the length of the ISI
    return isi * t_del #multiply by step size to get unit of time
    
print('inverse of ISI for 100ms:', ISI() ** -1)        
```

```python
t_max = 1000 #the stop time
t_del = 0.0005 #the integration step

int_potential = modified_euler(integrate_potential,E_m,t_max,t_del)

#plotting the membrane potential for 500ms-----

t = np.linspace(0,t_max*t_del,t_max) 
plt.figure(figsize=[12,8])
plt.plot(t,int_potential[0])
plt.title('membrane potential')
plt.ylabel('membrane potential (V)')
plt.xlabel('time (s)')
plt.grid()
plt.show()

print('firing rate for 500ms:',firing_rate())
print('inverse of ISI for 500ms:', ISI() ** -1) 
```

### c) Mean firing rate as a function of input current

```python
points = 500
rate = np.zeros(points)
isi = np.zeros(points)
I0 = np.linspace(0,5e-9,points)

for i in range(points):
    
    I_e = I0[i] #this is the spike count firing rate
    int_potential = modified_euler(integrate_potential,E_m,t_max,t_del)
    
    rate[i] = firing_rate() #spike count for the input current bounds
    isi[i] = ISI() #inter-spike interval for the input current bounds
    if isi[i] == 0:
        isi[i] = 1 #this is to avoid a divide-by-zero erro at very low current
```

```python
def theoretical_rate(I_e): #theoretical expression for firing rate
    num = (R_m * I_e) + E_m - V_res
    den = (R_m * I_e) + E_m - V_thr 
    
    #here the log argument is masked withe np.ma to avoid runtime errors
    rate = (tau_m * np.ma.log(num/den)) ** -1 
    
    return rate
```

```python
V_res = -0.08 #reset (V)
V_thr = -0.054 #threshold (V)

#plotting all of the ways to determine firing rate for different currents-----

plt.figure(figsize=[12,8])
plt.plot(I0,theoretical_rate(I0),label='analytical',c='k')
plt.plot(I0,rate,label='spike count')
plt.plot(I0,isi ** -1,label='inverse ISI')
plt.title('analytical firing rate vs. input current')
plt.ylabel('firing rate (spike # per second)')
plt.xlabel('input current (nA)')
plt.legend()
plt.grid()
plt.show()
```

After plotting all three methods of obtaining the firing rate, we can see that while both spike count and inverse ISI are close to the analytical value, they do vary slightly. The inverse ISI is closest to the analytical while the spike count method gives a value less than the "real" value. For both of the numerical methods (in the 500ms time range), the firing rate looks to be constant for a small range of input current before hitting a value of current that then raises the firing rate in a stair-step like fashion. 

I was also able to get the firing rates plotted for the 100ms time range as well, but unable to include both seperately due to time reasons. For the 100ms time range case in my code, the analytical soulution was being overshot by quite a large margin by both of the numerical values.

Interpreting the results, one can learn that as the input current is raised above a certain point (about 2.6nA for our case) the neuron is able to begin to fire. As the input current is raised further, the firing rate is increased as well. For a real neuron, the firing rate curve would be similar. However, on the top end of the current the firing rate would reach an asymptote as the cell saturates with current, whereas our models here look like they will continue to rise.
