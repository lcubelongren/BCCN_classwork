```python
x1a,x2a,yTa = training[:,0],training[:,1],training[:,2] #input variables x and target values
x1b,x2b,yTb = validation[:,0],validation[:,1],validation[:,2] #input variables x and target values
```

### a) Preprocessing:

```python
#starting point

plt.scatter(x1a,x2a,c=yTa)
plt.title('training data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
#find the center of the data

def center(x):
    
    x_center = (1/len(x)) * np.sum(x)
    
    return x_center

plt.scatter(x1a,x2a,c=yTa)
plt.scatter(center(x1a),center(x2a),color='r')
plt.title('training data w/ center')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
#now set the center to (0,0)

def centering(x):
    
    x_new = x - center(x)
    
    return x_new

xa_cen = [centering(x1a),centering(x2a)] #centered training input

plt.scatter(xa_cen[0],xa_cen[1],c=yTa)
plt.scatter(center(xa_cen[0]),center(xa_cen[1]),color='r') #center of centering, so (0,0)
plt.title('centered training data w/ center')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
#covariance, eigenvalues, and eigenvectors

def covariance(x1,x2): #covariance function of two inputs using numpy
    
    return np.cov(x1,x2,bias=True)

xa_cov = covariance(xa_cen[0],xa_cen[1]) #covariance of the training inputs
print('covariance: training')
print(xa_cov)

def eigen(x_cov): #eigen function of covariance using numpy
    
    w,v = np.linalg.eig(x_cov) #eigenvalues and normalized eigenvectors
    
    return w,v

eigen_a = eigen(xa_cov)

#eigenvalues of the training inputs in a diag. matrix
xa_val = np.array([[eigen_a[0][0],0],[0,eigen_a[0][1]]]) 
xa_vec = eigen_a[1] #eigenvectors of the training inputs
print('eigenvalues: training')
print(xa_val)
print('eigenvectors: training')
print(xa_vec)
```

```python
#decorrelation of the inputs

def decorrelation(eig_vec,x): #decorrelation function of eigenvectors and centered inputs
    
    dec = np.dot(eig_vec.T,x)
    
    return dec

xa_dec = decorrelation(xa_vec,xa_cen)

plt.scatter(xa_dec[0],xa_dec[1],c=yTa)
plt.title('decorrelated training data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
#sphering the inputs

def sphered(eig_val,x): #sphering function of eigenvalues and decorrelated inputs
    
    sph = np.dot(np.sqrt(inv(eig_val)),x)
    
    return sph

xa_sph = sphered(xa_val,xa_dec)

plt.scatter(xa_sph[0],xa_sph[1],c=yTa)
plt.title('sphered training data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
#do the same but for the validation set

def centering_different(x_validation,x_training): #new function so that we can have the same <x>
    
    x_new = x_validation - center(x_training)
    
    return x_new

xb_cen = [centering_different(x1b,x1a),centering_different(x2b,x2a)] #centered validation input

plt.scatter(xb_cen[0],xb_cen[1],c=yTb)
plt.title('centered validation data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
#the the same sphering transformation obtained from the 
#eigendecomposition of the centered training data’s covariance matrix

xb_dec = decorrelation(xa_vec,xb_cen)

plt.scatter(xb_dec[0],xb_dec[1],c=yTb)
plt.title('decorrelated validation data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

```python
xb_sph = sphered(xa_val,xb_dec)

plt.scatter(xb_sph[0],xb_sph[1],c=yTb)
plt.title('sphered validation data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

### b) Expand the sphered 2D input to all possible monomials up to degree 9

```python
def expansion(degree,x1,x2,N):
    
    phi = []
    
    for n in range(N):

        phi0 = np.zeros(shape=(degree+1,degree+1))
        
        for i in range(degree+1):
    
            for j in range(degree+1):
            
                if i + j <= degree:
            
                    phi0[i][j] = (x1[n] ** i) * (x2[n] ** j)
                
                else:
                    
                    phi0[i][j] = 0
                
        phi.append(phi0) 
            
    return np.array(phi)
```

```python
degree = 9 #9th degree monomials

def monomial(degree,x1,x2,N): #this function returns one value for each monomial-input pair

    value = expansion(degree,x1,x2,N)

    monomial = []
    
    for i in range(degree+1):
        
        for j in range(degree+1):
        
            if i + j <= degree:
    
                monomial.append(value[:,i,j])
        
    return np.array(monomial)
```

```python
x1_a = xa_sph[0] #training input 1
x2_a = xa_sph[1] #training input 2

N_a = len(x1_a) #number of inputs, x1=x2

p = 10 #plotting stuff
fig, axs = plt.subplots(1, p, figsize=(30, 5), sharey=True)
for i in range(p):

    axs[i].scatter(x1_a,x2_a,c=monomial(degree,x1_a,x2_a,N_a).T[:,i])
    axs[i].set_title('#%i' %(i+1))
    axs[i].set_ylabel('x1')
    axs[i].set_xlabel('x2')
    fig.suptitle('training set monomials for degree %i' %degree,fontsize=14,va='baseline')

plt.show()
```

```python
x1_b = xb_sph[0] #validation input 1
x2_b = xb_sph[1] #validation input 2

N_b = len(x1_b) #number of inputs, x1=x2

p = 10 #plotting stuff
fig, axs = plt.subplots(1, p, figsize=(30, 5), sharey=True)
for i in range(p):

    axs[i].scatter(x1_b,x2_b,c=monomial(degree,x1_b,x2_b,N_b).T[:,i])
    axs[i].set_title('#%i' %(i+1))
    axs[i].set_ylabel('x1')
    axs[i].set_xlabel('x2')
    fig.suptitle('validation set monomials for degree %i' %degree,fontsize=14,va='baseline')

plt.show()
```

```python
def predicted():
    
    monomial_training = monomial(degree,x1_a,x2_a,N_a) #training monomials
    monomial_validation = monomial(degree,x1_b,x2_b,N_b) #validation monomials
    
    #compute the weight in two parts
    w1 = inv(np.dot(monomial_training, monomial_training.T))
    
    w2 = w1.dot(monomial_training).dot(yTa.T)

    predicted = np.dot(w2,monomial_validation)
    
    return predicted
    
plt.scatter(x1_b,x2_b,c=predicted())
plt.title('predicted function y(x;w)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()
plt.show()
```

### c) Regularization using a weight-decay term

```python
#shuffle the data into 10 sets for cross-validation
def cross_validation(x1,x2,y):
    
    N = len(x1a) #size of data inputs
     
    #random int without replacement
    n1 = sample(range(N), N) 
    n2 = sample(range(N), N)
    
    #now lets plit the inputs into 10 folds
    
    fold_1 = [np.zeros(int(N/10)) for _ in range(10)] #emptiness
    fold_2 = [np.zeros(int(N/10)) for _ in range(10)] #emptiness
    known_y = [np.zeros(int(N/10)) for _ in range(10)] #emptiness
    
    count = 0
    for i in range(10):
        
        for j in range(int(N/10)):
            
            count += 1 #to get a new input each time
            
            number1 = n1[count-1] #random number
            fold_1[i][j] = x1[number1] #corresponding to random input

            number2 = n2[count-1]
            fold_2[i][j] = x2[number2]
            
            known_y[i][j] = y[number1] #and retain known output for later use

            
    return fold_1,fold_2,known_y
                   
#this gives us 10 arrays of randomly chosen numbers 
#without returning the same chosen number
```

```python
def predicted_new(lam,x1,x2,y,N): #predicted function but with lambda
    
    monomial_training = monomial(degree,x1,x2,N) #training monomials
    monomial_validation = monomial(degree,x1,x2,N) #validation monomials

    #compute the weight in two parts
    w1 = inv(np.dot(monomial_training, monomial_training.T) + lam) #here is the lambda term
    
    w2 = w1.dot(monomial_training).dot(y.T)

    predicted = np.dot(w2,monomial_validation)
    
    return predicted
```

```python
def MSE(y,y_known,N): #mean squared error function
    
    mse = (1/N) * np.sum((y - y_known) ** 2)
 
    return mse
```

```python
#define the regularization strengths
z = np.linspace(-4,4,81)
lam = 10**z #all the lambdas

def minimize():
    
    fold_1 = cross_validation(x1a,x2a,yTa)[0] 
    fold_2 = cross_validation(x1a,x2a,yTa)[1]
    known_y = cross_validation(x1a,x2a,yTa)[2]

    mse = np.zeros(len(lam))
        
    for j in range(len(lam)):
    
        mse0 = np.zeros(9)
    
        for i in range(9): #for each test set fold (#1-9, and then #10 is validation)
    
            y = predicted_new(lam[j],fold_1[i],fold_2[i],known_y[i],20)
        
            mse0[i] = MSE(y,known_y,len(y)) #mse0 is the error for an individual lambda

            mse[j] = np.sum(mse0) #mse is the error of all lambdas
            
    return mse

mse_mean = np.sum(minimize()) / len(lam)
print('mean mean-squared error:', mse_mean)
 
def sd(): #standard deviation
    
    sd = np.zeros(len(lam))
    
    for i in range(len(lam)):
    
        sd[i] = sqrt( (1/19) * (np.sum(minimize() - mse_mean)) ** 2)

    return sd
        
plt.plot(lam,minimize())
#plt.errorbar(lam, minimize(), sd(), linestyle='None', marker='^')
plt.title('sphered validation data')
plt.xlabel('lambda')
plt.ylabel('mean-squared error')
plt.xscale('log')
plt.grid()
plt.show()
```
