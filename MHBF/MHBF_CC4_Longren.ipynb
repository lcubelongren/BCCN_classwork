### 1. Have a look at the paper by Laing and Chow (2002). Read at least the abstract and the chapter $3. Reduced Description$, up to the definition of Equation (9).

### 2. Write a function $laing\_chow$ which implements eqs. (1-4) from Laing and Chow (2002) and which can be given to $scipy.integrate.odeint$ as its first argument. The parameters of the model should be additional arguments of $laing\_chow$ such that you can investigate the behavior of the model for different parameter settings. 


```python
def laing_chow(y, t, I1, I2):
    alpha, beta = 0.2, 0.4
    phi_a, tau_a = 0.4, 20
    g1, g2 = 1, 1
    u1, u2, a1, a2 = y  # initial conditions
    # differential equations, with respect to time
    du1 = -u1 + np.heaviside((alpha*u1*g1 - beta*u2*g2 - a1 + I1), 1)
    du2 = -u2 + np.heaviside((alpha*u2*g2 - beta*u1*g1 - a2 + I2), 1)
    da1 = (-a1 + phi_a * np.heaviside((alpha*u1*g1 - beta*u2*g2 - a1 + I1), 1)) / tau_a
    da2 = (-a2 + phi_a * np.heaviside((alpha*u2*g2 - beta*u1*g1 - a2 + I2), 1)) / tau_a
    return [du1, du2, da1, da2]
```

### Reproduce qualitatively Fig. 9 of Laing and Chow (2002) by integrating the differential equation defined by $laing\_chow$ using $odeint(laing\_chow, x_0, t, args=(I_1, I_2))$ where $x_0$ should contain the initial values $u_1 = 1, a_1 = 0.1, u_2 = 0, a_2 = 0.25$, $t$ should range from 0 to 500, $I$ are the inputs with $I_1 = 0.43, I_2 = 0.5$ and all other parameters of the model should be set by default to $α = 0.2, β = 0.4, φ = 0.4, τ = 20, g1 = g2 = 1$ in $laing\_chow$.

```python
time_range = np.linspace(0, 500, 5000)
integration = odeint(func=laing_chow, y0=(1, 0, 0.1, 0.25), t=time_range, args=(0.43, 0.5)).T
print(np.shape(integration))
```

### 3. Let us define population 1 as active if $u1 > u2$ and equivalently for population 2. Determine the dominance durations $D_1$ during which population 1 is active, from your simulations. To do this write a function $dominance\_durations(x_1, x_2, t)$ where $x_1$ and $x_2$ are arrays containing the simulated states and $t$ is an array of the same length containing the corresponding times. 

```python
def dominance_durations(x1, x2, t):
    points = len(t)
    dt = np.max(t) / points
    dominance_1 = 0  # dominant population
    dominance_2 = 0  # non-dominant population
    switches = 0  # number of times dominance changes
    for i in range(points-1):
        # update dominance duration
        if x1[i] > x2[i]:  # population 1 active
            dominance_1 += dt
        else:  # population 2 active
            dominance_2 += dt
        # update switch number
        if x1[i] > x2[i] and x1[i-1] < x2[i-1]:
            switches += 1
        elif x1[i] < x2[i] and x1[i-1] > x2[i-1]:
            switches += 1
        else:
            pass
    return [dominance_1, dominance_2, switches]
```

```python
def analytical_dominance(I1, I2, beta, phi_a, tau_a):  # eq. 9
    return -tau_a * np.log((I2-beta)/(beta+phi_a-I1))
```

```python
D1 = dominance_durations(u1, u2, time_range)
D1_duration = D1[0]*2 / D1[2]
T1 = analytical_dominance(I1=0.43, I2=0.5, beta=0.4, phi_a=0.4, tau_a=20)
print('%.2f, dominance duration mean D1' %D1_duration)
print('%.2f, dominance duration T1' %T1)
```

### 2. Why can’t you use scipy.integrate.odeint to implement this model?

We cannot continue to use $odeint$ as we are now working with a stochastic differential equation that is not deterministic. Ordinary differential equations are handled well for the most part by $odeint$, but as the stochastic mean-field firing rate model in Moreno-Bote is not ordinary, we will need a different method of solving the differential equations.

### 3. Write a function $euler\_maruyama(ffun, gfun, x_0, t, $**$args)$ which implements this equation and which behaves like $odeint$. Note that we have defined args as a dictionary which differs from $odeint$ where args is a tuple. This makes it easier to change individual parameters of the model functions later. 

## <font color='black'> $ X_t = X_{t−1} + f(X_{t−1})\Delta t + g(X_{t−1})\sqrt{\Delta t} W_t $

```python
def euler_maruyama(ffun, gfun, x0, t, **kwargs):
    time_points = len(t)  # size of the time range
    dt = np.max(t)/time_points  # step size
    x = np.zeros((len(x0), time_points))  # empty output array
    W = np.random.normal(loc=0.0, scale=np.sqrt(dt), size=time_points)  # sample from a normal dist.
    x[:, 0] = x0
    g = lambda x: gfun  # defining the second function
    for i in range(time_points-1):  # perform the integration at each time point
        for n in range(len(x0)):  # for each equation
            x[n, i+1] = x[n, i] + (ffun(x, t, **kwargs)[n][i] * dt) + (g(x) * np.sqrt(dt) * W[i]) 
    return x
```

### 5. Finally, write functions $f\_moreno\_bote$ and $g\_moreno\_bote$ which implement equations (A1, B5, B6, B7) of Moreno-Bote et al. (2007). (hint: note the stochastic part of the equations: only that goes into $g\_moreno\_bote$.) 
### The equations B5 and B7 describe the dynamics of a single population $A$. You need to simulate a second population $B$ that follows the same dynamics (see Fig. 3B of the main text).

## <font color='black'> $ \frac{d}{dt} n = -\frac{n}{\tau_s} + \sigma \sqrt{\frac{2}{\tau_s}} \xi (t) \ \ \ \ \ (A1) $

## <font color='black'> $ \tau \frac{d}{dt} r_A = -r_A + f(\alpha r_A - \beta r_{A,inh} + g_A - \alpha_A + n_A) $

## <font color='black'> $ \tau_a \frac{d}{dt} \alpha_A = -\alpha_A + \gamma r_A \ \ \ \ \ (B5) $

## <font color='black'> $ f(x) = \langle 1 + exp\{-(x-\theta)/k\} \rangle ^{-1} \ \ \ \ \ (B6) $

## <font color='black'> $ r_{A,inh} = (r_{pool} + \eta r_A ) ^2 \ \ \ \ \ (B7) $

## <font color='black'> where $ r_{pool} = [\phi (r_A + r_B) + g_A + g_B]^+ $

```python
def f_moreno_bote(y, t, **args):  # the non-stochastic part
    r_A, r_B, alpha_A, alpha_B, n_A, n_B  = y  # initial conditions
    f = lambda x: 1 / np.mean(1 + np.exp(-(x - args['theta']) / args['k']))  # sigmoid function
    rect = lambda x: x if x >= 0 else 0  # rectification function
    r_pool = rect(args['phi'] * (r_A + r_B) + args['g_A'] + args['g_B'])
    r_Ainh = (r_pool + args['eta'] * r_A) ** 2
    r_Binh = (r_pool + args['eta'] * r_B) ** 2
    # differential equation, with respect to time
    dn_A = -n_A / args['tau_s']
    dn_B = -n_B / args['tau_s']
    dalpha_A = (-alpha_A + args['gamma'] * r_A) / args['tau_a']
    dalpha_B = (-alpha_B + args['gamma'] * r_B) / args['tau_b']
    dr_A = (-r_A + f(args['alpha'] * r_A - args['beta'] * r_Ainh + args['g_A'] - alpha_A + n_A)) / args['tau']
    dr_B = (-r_B + f(args['alpha'] * r_B - args['beta'] * r_Binh + args['g_B'] - alpha_B + n_B) ) / args['tau']
    return [dr_A, dr_B, dalpha_A, dalpha_B, dn_A, dn_B]
```

```python
def g_moreno_bote(y, **args):  # the stochastic part
    xi = np.random.normal(loc=0, scale=0.03, size=1)  # white noise
    # differential equation, with respect to time
    dn = args['sigma'] * np.sqrt(2 / args['tau_s'])# * xi
    return [0, 0, 0, 0, dn, dn]
```

### Plot a histogram of the dominance durations. What is their mean?

```python
D_AB = dominance_durations(r_A, r_B, time_range)
D_A_duration = D_AB[0] / D_AB[2]
D_B_duration = D_AB[1] / D_AB[2]
print('%.2f, dominance duration mean of population A' %D_A_duration)
print('%.2f, dominance duration mean of population B' %D_B_duration)
```

```python
# modifying our previous function to enable histogram plotting
def histogram_dominance(x1, x2, t):
    points = len(t)
    dt = np.max(t) / points
    # keep track of the order and point where switching occurs
    dominance_1 = []
    dominance_2 = []
    switch_point = []
    for i in range(points-1):
        # update dominance duration
        if x1[i] > x2[i]:  # population 1 active
            dominance_1 = np.append(1, dominance_1)
            dominance_2 = np.append(dominance_2, 0)
        else:  # population 2 active
            dominance_2 = np.append(dominance_2, 1)
            dominance_1 = np.append(dominance_1, 0)
        # update switch number
        if x1[i] > x2[i] and x1[i-1] < x2[i-1]:
            switch_point = np.append(switch_point, i)
        elif x1[i] < x2[i] and x1[i-1] > x2[i-1]:
            switch_point = np.append(switch_point, i)
        else:
            pass
    # histogram friendly part
    D_A_duration = []
    D_B_duration = []
    for i in range(1, len(switch_point)):
        if dominance_1[int(switch_point[i]-1)] == 1:
            D_A_duration = np.append(D_A_duration, switch_point[i]-switch_point[i-1])
        elif dominance_2[int(switch_point[i]-1)] == 1:
            D_B_duration = np.append(D_B_duration, switch_point[i]-switch_point[i-1])
        else:
            pass
    return [np.dot(D_A_duration, dt), np.dot(D_B_duration, dt)]
```

```python
hist_dominance = histogram_dominance(r_A, r_B, time_range) 
plt.hist(hist_dominance[0], histtype='stepfilled', bins=100, label='Dominance Duration A')
plt.hist(hist_dominance[1], histtype='stepfilled', bins=100 ,label='Dominance Duration B')
plt.ylabel('number of occurances')
plt.xlabel('dominance duration')
plt.legend()
plt.show()
```

```python
D_A_mean = np.mean(hist_dominance[0])
D_B_mean = np.mean(hist_dominance[1])
print('%.2f, dominance duration mean of population A' %D_A_mean)
print('%.2f, dominance duration mean of population B' %D_B_mean)
```
