## Exercise 1: Chapman-Kolmogorov

### 1. Since we are interested in the time evolution of the distribution, you should first generate an $N × M$ matrix where the first column contains an initial condition ensuring the particle rests in the center bin. Use $N = 2 · 50 + 1$ dimensionless positions (each a distance of 1 apart, so $∆x = 1$) and $M = 1000/∆t$, where $∆t = 1$.

```python
dt = 1
N = 2 * 50 + 1
M = 1000 // dt
time_matrix_CK = np.zeros((N, M))  # time evolution for each position
time_matrix_CK[int(N/2+1), 0] = 1  # 100% probability of resting in center bin
print(np.shape(time_matrix_CK), ' N , M')
plt.imshow(time_matrix_CK[:,:10], origin='lower', aspect='auto')
plt.title('Initial Time Matrix')
plt.xlabel('time')
plt.ylabel('position')
plt.colorbar(label='probability')
plt.show()
```

### 2. Use the transition probability,

## <font color='black'> $ P(x|x';\Delta t) = \frac{1}{\sqrt{2 \pi \sigma^2 \Delta t}} exp \{ -\frac{(x - x' - \mu \Delta t)^2}{2 \sigma^2 \Delta t} \} $

### to compute the matrix of probabilities of transitioning to any $x$ from any $x_0$ (take $µ = 0.1$, $σ = 1.0$). 
### This should result in an $N × N$ matrix where each row contains the probabilities of transition from a particular $x_0$ (the columns) into the corresponding $x$ (the row) in the next time step.

```python
mu, sigma = 0.1, 1.0
print('mu:', mu, 'sigma:', sigma)
x_values = np.linspace(1, N, N, dtype=int)
trans_matrix_CK = np.zeros((N, N))  # transition probability matrix
Trans_Prob = lambda x, x_prime: (1/np.sqrt(2*np.pi*(sigma**2)*dt)) * np.exp((-(x-x_prime-mu*dt)**2)/(2*(sigma**2)*dt))
for i in x_values:  # loop over rows (x)
    for j in x_values:  # loop over columns (x_0)
        trans_matrix_CK[i-1,j-1] = Trans_Prob(x=i, x_prime=j)
print(np.shape(trans_matrix_CK), 'x_0 , x')
plt.imshow(trans_matrix_CK, origin='lower')
plt.title('Transition Probability - Chapman-Kolmogorov')
plt.xlabel('$x_0$')
plt.ylabel('$x$')
plt.colorbar(label='probability')
plt.show()
```

## Exercise 2: Fokker-Planck

### 1. Once again create a matrix for the distribution over time as per 1.1.

```python
dt = 1
N = 2 * 50 + 1
M = 1000 // dt
time_matrix_FP = np.zeros((N, M))  # time evolution for each position
time_matrix_FP[int(N/2+1), 0] = 1  # 100% probability of resting in center bin
print(np.shape(time_matrix_FP), ' N , M')
```

### 2. Instead of using the Chapman-Kolmogorov equation to determine the transition probabilities, 
### this time we will integrate the Fokker-Planck equation that results from the transition probability,

## <font color='black'> $ \frac{\partial P(x,t|x_0)}{\partial t} = -\mu \frac{\partial P}{\partial x} + \frac{\sigma^2}{2}\frac{ \partial^2 P}{\partial x} $ ,

### If one performs a discrete integration using Euler the distribution evolves according to:

## <font color='black'> $ P(x,t+\Delta t) = (1-r\sigma^2) P(x,t) + \frac{1}{2} (-\Delta x r \mu + r\sigma^2) P(x+\Delta x, t) + \frac{1}{2} (\Delta x r \mu + r \sigma^2) P(x-\Delta x, t) $ 

### where $r = \Delta t / \Delta x^2$. Create an $N × N$ matrix that implements this discretized Fokker-Planck differential operator.

```python
def FP_Euler(dx, dt, sigma, mu, N):
    r = dt / (dx**2)
    P = np.zeros((N,N))
    for i in range(N):  # loop over the diagonal
        P[i,i] = (1 - r*(sigma**2))
        if i != N-1:
            P[i,i+dx] = 0.5 * (-dx*r*mu + r*(sigma**2))
        if i != 0:
            P[i,i-dx] = 0.5 * (dx*r*mu + r*(sigma**2))
    return P  # NxN transition matrix
```

```python
trans_matrix_FP = FP_Euler(dx=1, dt=1, sigma=1.0, mu=0.1, N=N)
print(np.shape(trans_matrix_FP), 'x_0 , x')
plt.imshow(trans_matrix_FP, origin='lower')
plt.title('Transition Probability - Fokker-Planck, $\Delta t = 1$')
plt.xlabel('$x_0$')
plt.ylabel('$x$')
plt.colorbar(label='probability')
plt.show()
```

### 3. Use the transition matrix from the previous step to compute the time-dependent distribution as in Exercise 1. 
### Plot the result. Look closely at the first several time steps, and comment on whether the system behaves as you expect.

```python
np.random.seed(111)
time_points = np.shape(time_matrix_FP)[1]  # number of time steps
time_matrix_FP = np.zeros((N, M))  # time evolution for each position
time_matrix_FP[int(N/2+1), 0] = 1  # 100% probability of resting in center bin
for t in range(1, time_points):  # loop over time
    time_matrix_FP[:,t] = np.dot(time_matrix_FP[:, t-1], trans_matrix_FP)
print(np.shape(time_matrix_FP), 'N , M')
plt.imshow(time_matrix_FP, origin='upper', aspect='auto', vmin=0, vmax=0.05)
plt.title('Time Evolution - Fokker-Planck, $\Delta t = 1$')
plt.xlabel('time')
plt.ylabel('position')
cbar = plt.colorbar(ticks=[0, 0.025, 0.05], label='probability')
cbar.ax.set_yticklabels(['0%', '2.5%', '> 5%'])
plt.show()

plt.imshow(time_matrix_FP[:70,:350], origin='upper', aspect='auto', vmin=0, vmax=0.1)
plt.title('Zoomed In - Fokker-Planck, $\Delta t = 1$')
plt.xlabel('time')
plt.ylabel('position')
cbar = plt.colorbar(ticks=[0, 0.05, 0.1], label='probability')
cbar.ax.set_yticklabels(['0%', '5%', '> 10%'])
plt.show()
```

Due to our transition matrix with $\Delta t = 1$, this result makes sense. Starting in time bin zero, each time step is sampling from the previous distribution.  
As our transition matrix is zero for anything outside of the three diagonal points, the evolution over time is limited to the three points surrounding the initial central bin.  
Also, due to our choice in $\Delta t$, the time evolution is a kind of "checker board" pattern as the middle diagonal point is zero.  

```python
color = iter(cm.viridis(np.linspace(0, 1, 11)))
[plt.plot(time_matrix_FP[:,5*i], color=next(color), label='$t=%i$' %(i*5)) for i in range(11)]
plt.title('Time Evolution - Fokker-Planck, $\Delta t = 1$')
plt.xlabel('position')
plt.legend()
plt.grid()
plt.show()
```

This plot reaffirms our previous thought. For $t \ne 0$, the probability fluctuates between zero and non-zero over space.

### 4. Recompute the $N × N$ transition matrix but with $∆t = 0.5$ (you’ll also need a new $N × M$ matrix for the distribution), and see if that improves on the previous result. Can you guess why this might be?

```python
np.random.seed(111)
trans_matrix_FP_half = FP_Euler(dx=1, dt=0.5, sigma=1.0, mu=0.1, N=N)
plt.imshow(trans_matrix_FP_half, origin='lower')
plt.title('Transition Probability - Fokker-Planck, $\Delta t = 0.5$')
plt.xlabel('$x_0$')
plt.ylabel('$x$')
plt.colorbar(label='probability')
plt.show()

time_matrix_FP_half = np.zeros((N, M))  # time evolution for each position
time_points_half = np.shape(time_matrix_FP_half)[1]  # number of time steps
time_matrix_FP_half[int(N/2+1), 0] = 1  # 100% probability of resting in center bin
for t in range(1, time_points):  # loop over time
    time_matrix_FP_half[:,t] = np.dot(time_matrix_FP_half[:, t-1], trans_matrix_FP_half)
plt.imshow(time_matrix_FP_half, origin='upper', aspect='auto', vmin=0, vmax=0.05)
plt.title('Time Evolution - Fokker-Planck, $\Delta t = 0.5$')
plt.xlabel('time')
plt.ylabel('position')
cbar = plt.colorbar(ticks=[0, 0.025, 0.05], label='probability')
cbar.ax.set_yticklabels(['0%', '2.5%', '> 5%'])
plt.show()
```

Changing the value of $\Delta t$ to 0.5 leads to the time evolution of FP that now does not have the previous pattern.   
This stems from the new transition matrix, which now has three non-zero values along the diagonal.  

```python
color = iter(cm.viridis(np.linspace(0, 1, 11)))
[plt.plot(time_matrix_FP_half[:,5*i], color=next(color), label='$t=%i$' %(i*5)) for i in range(11)]
plt.title('Time Evolution - Fokker-Planck, $\Delta t = 0.5$')
plt.xlabel('position')
plt.legend()
plt.grid()
plt.show()
```

### 5. Plot $−1/ log(P(t) + \epsilon)$ for a few early time steps. Make the same plots for the Chapman-Kolmogorov solution. 
### How are they different? Can you explain why?

```python
epsilon = 0.1
Log_Function = lambda P, epsilon: -1 / np.log(P + epsilon)
log_FP = Log_Function(time_matrix_FP_half, epsilon)
log_CK = Log_Function(time_matrix_CK, epsilon)

fig, axs = plt.subplots(2,1, sharex=True, sharey=True)
color = iter(cm.viridis(np.linspace(0, 1, 9)))
[axs[0].plot(log_FP[:,25*i], color=next(color), label='$t=%i$' %(i*25)) for i in range(1,10)]
axs[0].set_title(r'$-1 / log(P(t) + \epsilon)$, $\epsilon = %.1f$ - Fokker-Planck, $\Delta t = 0.5$' %epsilon)
axs[0].legend()
axs[0].grid()

color = iter(cm.viridis(np.linspace(0, 1, 9)))
[plt.plot(log_CK[:,25*i], color=next(color), label='$t=%i$' %(i*25)) for i in range(1,10)]
axs[1].set_title(r'$-1 / log(P(t) + \epsilon)$, $\epsilon = %.1f$ - Chapman-Kolmogorov' %epsilon)
axs[1].set_xlabel('position')
axs[1].legend()
axs[1].grid()
plt.tight_layout()
plt.show()
```

Comparing this negative inverse log function of the two probability evolutions, they are a bit different.  
For the Fokker-Planck, each time point's peak moves more slowly away from the center bin than the Chapman-Kolmogorov method.

### 8. Why does the initial condition have this effect on the computed distributions? Look at the derivation of the Fokker-Planck equation from the lecture and pay close attention to which step could fail for the first initial condition.

The initial conditon has a very large impact on the computed time evolution distribuition, as the transition matrix is limiting for the Fokker-Planck method.  

From the lecture notes, we can see why this is:  
*In this case, the transition probability $p(x|x_0; ∆t)$ very quickly decays as a function of the difference $|x − x_0|$, because in small intervals, the decision variable cannot travel along way.*
Therefore, the two initial conditions we have looked at do have a large impact on the evolution of the distribution.  
The difference between each time step for our first condition (probability of 1 in the central bin) was very large, and thus had a huge impact. 
For the second condition the difference was smaller, and therefore the distribution was less "choppy" over time.
