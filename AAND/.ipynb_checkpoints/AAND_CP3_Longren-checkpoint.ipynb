### (c) Compute the coefficient of variation CV from the pooled ISIs:

The coefficient of variation,

## <font color='black'> $ CV = \frac{\sigma_{\tau}}{\langle \tau \rangle} $

where $_$ and $ \langle  \rangle $ are the standard deviation and the mean of the inter-spike intervals $$ (ISIs).

```python
def CV(tau):
    tau = tau[~np.isnan(tau)]  # remove nan values
    std_dev = np.std(tau)  # standard deviation
    mean = np.mean(tau)  # mean
    return std_dev / mean
```

```python
print('Coefficient of variation:', CV(ISI), ', from pooled ISIs')
```

### (d) Compute and plot both the trial-resolved and the time-resolved Fano factor F in non-overlapping windows of lengths $T_W  [0.3,0.5,1] \ s$. Interpret the results. How does the window size affect the Fano factor in this dataset?

The Fano factor,

## <font color='black'> $ F = \frac{\sigma_n^2}{\langle n \rangle} $

where $_n^2$ and $\langle n \rangle$ are the variance and the mean of the spike counts $n$ in time windows of length $T_W$.

```python
def Fano(data, period, T_W, resolve):
    # data = data[~np.isnan(data)]  # remove nan values
    # determine the spike counts over the time window, both trial-resolved and time-resolved
    # dimensions are: (trial resolved: trial #, time resolved: time window #)
    
    window_num = int(period // T_W)
    # the following if statements are to make this code work with part #2 and #3
    if type(data[0]) == np.float64:
        trial_num = 1
    else:
        trial_num = len(data[0])
    if np.shape(data) == (1000,):
        spike_num = 1000
    else:
        spike_num = len(data[:,0])
    
    
    # below, I attempt to calculate the Fano factor both ways:
    
    if resolve == 'trial':  # average over trials
        F = np.zeros(trial_num)
        for t in range(trial_num):  # loop over trials
            n = np.empty(0)  # empty array to count
            for i in range(spike_num):  # loop over spikes
                for j in range(window_num):  # loop over windows
                    if (T_W*j) <= data[i,t] < (T_W*j + T_W):  # if data is in the window
                        n = np.append(n, [data[i,t]])  # add to the n array
                    else:
                        pass
            F[t] = np.var(n) / np.mean(n)  # calculate Fano factor
        
    elif resolve == 'time':  # average over time
        F = np.zeros(window_num)
        for w in range(window_num):  # loop over windows
            n = np.empty(0)  # empty array to count
            for i in range(spike_num):  # loop over spikes
                for j in range(trial_num):  # loop over trials
                    if (T_W*w) <= data[i,j] < (T_W*w + T_W):  # if data is in the window
                        n = np.append(n, [data[i,j]])  # add to the n array
                    else:
                         pass
            F[w] = np.var(n) / np.mean(n)  # calculate Fano factor
            
    else:
        return 'resolve type not understood'
    
    return F
```

```python
T_W = [0.3, 0.5, 1]  # window lengths [s]
period = 5.5  # trial length [s]

for T in T_W:  # plot the trial resolved Fano factors
    fano = Fano(data=spike_times_seconds, period=period, T_W=T, resolve='trial')
    trial_num = len(spike_times_seconds[0,:])
    trials = np.linspace(1, trial_num, trial_num)
    plt.plot(trials, fano)
    plt.scatter(trials, fano, label='Window Length = %s s' %T)
    plt.title('Trial Resolved Fano Factors')
    plt.ylabel('Fano factor')
    plt.xlabel('trial #')
    plt.legend()
    plt.grid()
plt.show()

for T in T_W:  # plot the time resolved Fano factors
    fano = Fano(data=spike_times_seconds, period=period, T_W=T, resolve='time')
    window_num = int(period/T)
    windows = np.linspace(1, window_num, window_num)
    plt.plot(windows, fano)
    plt.scatter(windows, fano, label='Window Length = %s s' %T)
    plt.title('Time Resolved Fano Factors')
    plt.ylabel('Fano factor')
    plt.xlabel('time window #')
    plt.legend()
    plt.grid()
plt.show()
```

### (c) Compute and plot the spike-train auto-correlation histogram by discretizing the auto-correlation function.

## <font color='k'> $ Q_{pp} (\tau) = \frac{1}{T} \int_0^T dt \  \langle \rho (t) \rho (t+\tau) \rangle - \langle r \rangle ^2 $

### Use bin size $t = 0.1 ms$ and multiply the result by $t$ such that the auto-correlation in zero equals the average firing rate $\langle r \rangle$.
###  Plot the auto-correlation both with and without the central peak. 
### Also try different x-axis limits to see if you can observe any interesting structure.

```python
def auto_correlation(data1, data2, mode):
    result = np.correlate(data1[0], data2[0], mode)
    return result[result.size//2:]  # take the positive values, for the interal bounds
```

```python
bin_size = 0.1  # [ms]
period = 100  # [ms]

acn = auto_correlation(spike_times_hom, spike_times_inh, 'full') * bin_size
plt.hist(acn, bins=int(period/bin_size), histtype='step', color='k')
plt.ylabel('auto-correlation')
plt.xlabel('time [ms]')
plt.show()
```
