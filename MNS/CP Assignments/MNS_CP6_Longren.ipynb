## 1. Potassium channel

### a) Write functions defining the voltage-dependent transition rates

```python
def alpha_n(V):
    num = 0.01 * (V+55)
    den = 1 - np.exp(-0.1 * (V+55))
    return num/den
```

```python
def beta_n(V):
    beta = 0.125 * np.exp(-0.0125 * (V+65))
    return beta
```

We will also rewrite $dn/dt$ for use later.

```python
#probability of opening and closing of an ion gate
def n_derivative(n,t,alpha_n,beta_n): 
    der = alpha_n * (1-n) - beta_n * n
    return der
```

### b) Plot the activation time constant & steady-state activation

```python
def time_constant(alpha_n,beta_n):
    tau = (alpha_n + beta_n) ** -1
    return tau
```

```python
def steady_activation(alpha_n,beta_n):
    n_inf = alpha_n * (alpha_n + beta_n) ** -1
    return n_inf
```

```python
#plot time constant
def plot_tc(points,vspan): 
    V_span = np.linspace(-vspan,vspan,points) #empty voltage
    t_constant = np.zeros(points) #empty time constant
    for i in range(points):
        alpha = alpha_n(V_span[i])
        beta = beta_n(V_span[i])
        t_constant[i] = time_constant(alpha,beta)
    plt.plot(V_span,t_constant,label='time constant')
```

```python
#plot steady-state activation
def plot_sa(points,vspan): 
    V_span = np.linspace(-vspan,vspan,points) #empty voltage
    s_activation = np.zeros(points) #empty steady-state
    for i in range(points):
        alpha = alpha_n(V_span[i])
        beta = beta_n(V_span[i])
        s_activation[i] = steady_activation(alpha,beta)
    plt.plot(V_span,s_activation,label='steady-state')
    plt.scatter(-40,s_activation[110],c='k') #the value at -40mV
    print('n_inf(-40mV) =',s_activation[110])
```

```python
plt.figure(figsize=(10,6))
plot_tc(300,150) #plot time constant over 300 data points
plt.title('Potassium Time Constant vs. Potential')
plt.ylabel(r'time constant $\tau_n$ (ms)')
plt.xlabel('voltage (mV)')
plt.grid()

plt.figure(figsize=(10,6))
plot_sa(300,150) #plot steady-state over 300 data points
plt.title('Potassium Steady-State vs. Potential')
plt.ylabel(r'steady-state $n_\infty$')
plt.xlabel('voltage (mV)')
plt.grid()

plt.show()
```

### Verify the steady-state activation by looking at the transient response

```python
#euler function to solve ODEs
def euler(fx,x0,t_max,dt,parameters={}): 
    time = np.linspace(0,t_max*dt,t_max) #time range
    x = np.zeros(t_max) #emptiness
    x[0] = x0 #initial condition

    for i in range(t_max-1): #two-dimensional
        
        x[i+1] = x[i] + fx(x[i],time[i],*parameters) * dt
        
    return x,time
```

```python
#plot the transient response for a single voltage over time
def plot_tr(voltage,x0,t_max,dt): 
    alpha = alpha_n(voltage)
    beta = beta_n(voltage)
    
    #lambda defines a function, where here it is a function of n & t
    #this allows us to use the n_derivative function with the euler function
    n_derivative2 = lambda n, t: n_derivative(n,t,alpha,beta)
    
    transient_response = euler(n_derivative2,x0,t_max,dt)[0] #the variable n
    time = euler(n_derivative2,x0,t_max,dt)[1] #time axis
    plt.plot(time,transient_response)
    plt.scatter(23,transient_response[2300],c='k')
    print('steady-state @ 23ms =',transient_response[2300])
```

```python
plt.figure(figsize=(10,6))
plot_tr(-40,0,2500,0.01)
plt.title('Potassium Transient Response for $t_0$ = 0 & $V_c$ = -40 mV')
plt.ylabel(r'steady-state $n_\infty$')
plt.xlabel('time (ms)')
plt.grid()
plt.show()
```

### c) Simulate the current responses to voltage steps under voltage-clamp conditions

```python
#define function for voltage-clamp conditions
def voltage_clamp(t,V_c,t_change): 
    if t >= t_change: 
        V = V_c #command voltage
    else:
        V = -65 #(mV)
    return V
```

```python
def n_derivative_clamp(n,t,V_c,t_change): 
    if t >= t_change: 
        V = V_c #command voltage
    else:
        V = -65 #(mV)
    der = alpha_n(V) * (1-n) - beta_n(V) * n
    return der
```

```python
#potassium current in the Hodgkin-Huxley model
def current_k(n,g_k,V,E_k):
    current_k = g_k * (n**4) * (V - E_k)
    return current_k
```

```python
#now to simulate the current responses to this voltage
def plot_ik(n0,V_c,t_max,dt): #plot the current responses
    
    g_k = 36e-6 #conductance potassium (mS)
    E_k = -77 #reversal potential potassium (mV)
    t_change = 2 #time condition (ms)

    time = np.linspace(0,t_max*dt,t_max) #time range
    
    #here we calculate the transient responses n over time
    n_derivative_clamp2 = lambda n, t: n_derivative_clamp(n,t,V_c,t_change)
    transient_response = euler(n_derivative_clamp2,n0,t_max,dt)[0]
    
    #here we calculate the current responses I over time
    voltage = np.linspace(0,t_max*dt,t_max)
    for i in range(t_max): 
        voltage[i] = voltage_clamp(time[i],V_c,t_change)
    current_response = current_k(transient_response,g_k,voltage,E_k)
    
    plt.plot(time,current_response,label='%i mV' %V_c)
```

```python
plt.figure(figsize=(10,6))
[plot_ik(0.3177,-10*i,4000,0.005) for i in range(4,11)]
plt.title('Potassium Current Responses $I_k$ under 7 Various Voltage-Clamp Conditions')
plt.ylabel('current response $I_K$ (mA)')
plt.xlabel('time (ms)')
plt.legend()
plt.grid()
plt.show()
```

Looking at the results of the potassium current responses under various voltage-clamp conditions, we can see that changing the value for $V_c$ has quite a large impact on the resulting current response. Between the voltages $-70mV$ and $-60mV$, there is a turning point where after $2ms$ the current response goes in a positive direction, rather than a negative current as is for lower clamped voltages.

From the previous time constant graph, we see a peak time constant at around $-80mV$. Also, the inflection point of the steady-state graph is a little under $-50mV$. At these points there is a noticable impact as for these voltages changes occur in the current response graph.

### d) Plot the current-voltage relation for the instantaneous and the steady-state potassium current

```python
def plot_iv(n,vspan,points,lbl): #plot the current responses
    
    g_k = 36e-6 #conductance potassium (mS)
    E_k = -77 #reversal potential potassium (mV)

    voltage = np.linspace(-vspan,vspan,points) #empty voltage
    current = current_k(n,g_k,voltage,E_k) #current calculation
    
    plt.plot(voltage,current,label=lbl)
```

```python
def plot_iv_steady(vspan,points,lbl): 
    voltage = np.linspace(-vspan,vspan,points) #empty voltage
    s_activation = np.zeros(points) #empty steady-state
    for i in range(points):
        alpha = alpha_n(voltage[i])
        beta = beta_n(voltage[i])
        s_activation[i] = steady_activation(alpha,beta)
    plt.plot(voltage,s_activation,label=lbl)
```

## 2. Sodium ion channel

### a) Write all necessary functions

```python
def current_na(m,h,V,g_na,E_na):
    current_na = g_na * (m**3) * h * (V - E_na)
    return current_na
```

```python
def alpha_m(V):
    alpha = 0.1 * (V + 40) / (1 - exp(-0.1 * (V + 40)))
    return alpha
```

```python
def alpha_h(V):
    alpha = 0.07 * exp(-0.05 * (V + 65))
    return alpha
```

```python
def beta_m(V):
    beta = 4 * exp(-0.0556 * (V + 65))
    return beta
```

```python
def beta_h(V):
    beta = 1 / (1 + exp(-0.1 * (V + 35)))
    return beta
```

### Plot the steady-state activation and time constant

```python
def plot_sodium_time(gating_alpha,gating_beta,vspan,points,lbl): #plot sodium time constant
    V_span = np.linspace(-vspan,vspan,points) #empty voltage
    t_constant = np.zeros(points) #empty time constant
    for i in range(points):
        alpha = gating_alpha(V_span[i])
        beta = gating_beta(V_span[i])
        t_constant[i] = time_constant(alpha,beta)
    plt.plot(V_span,t_constant,label=lbl)
```

```python
def plot_sodium_steady(gating_alpha,gating_beta,vspan,points,lbl): #plot sodium time constant
    V_span = np.linspace(-vspan,vspan,points) #empty voltage
    s_activation = np.zeros(points) #empty time constant
    for i in range(points):
        alpha = gating_alpha(V_span[i])
        beta = gating_beta(V_span[i])
        s_activation[i] = steady_activation(alpha,beta)
    plt.plot(V_span,s_activation,label=lbl)
```

Compared to the results of the $n$ variable from the potassium section, we can see that $m$ matches what $n$ was as they both move in a positive direction. Compared to the $h$ variable, which moves in the opposite direction for increasing voltage.

The steady-state $h$ is activated by depolarization, while the variable $m$ is deactivated by depolarisation (low voltage returning high state means activated). From the time constant graph, we can see that $h$ is fast and $m$ is slow.

### b)  Simulate and plot the sodium current

```python
def plot_is(m0,h0,V,t_max,dt): #plot the current responses for sodium
    
    g_na = 120e-6 #conductance sodium (mS)
    E_na = 50 #reversal potential sodium (mV)

    time = np.linspace(0,t_max*dt,t_max) #time range
    
    #transient response for m
    transient_m = lambda m, t: n_derivative(m,t,alpha_m(V),beta_m(V))
    m = euler(transient_m,m0,t_max,dt)[0]
    
    #transient response for h
    transient_h = lambda h, t: n_derivative(h,t,alpha_h(V),beta_h(V))
    h = euler(transient_h,m0,t_max,dt)[0]
    
    current_response = current_na(m,h,V,g_na,E_na)
    
    minimum = np.ndarray.min(current_response)
    print('minimum current (pA):',minimum * 1e7)
    print('steady-state @ 15ms (pA) =',current_response[1500] * 1e6)
    
    plt.plot(time,current_response,label='$V_c$ = %i mV' %V)
    plt.scatter(15,current_response[1500],c='k')
```

```python
plt.figure(figsize=(10,6))
plot_is(0.0529,0.5961,-20,3500,0.005)
plt.title('Sodium Current vs. Time')
plt.ylabel('current $I_{Na}$ (mA)')
plt.xlabel('time (ms)')
plt.legend()
plt.grid()
plt.show()
```

### c) Simulate the voltage clamp experiment for the sodium current

```python
def derivative_clamp(x,t,V_c,t_change,alpha,beta): 
    if t >= t_change: 
        V = V_c #command voltage
    else:
        V = -65 #(mV)
    der = alpha(V) * (1-x) - beta(V) * x
    return der
```

```python
#now to simulate the current responses to this voltage
def plot_is_clamped(m0,h0,V_c,t_max,dt): #plot the current responses
    
    g_na = 120e-6 #conductance sodium (mS)
    E_na = 50 #reversal potential sodium (mV)
    t_change = 2 #time condition (ms)

    time = np.linspace(0,t_max*dt,t_max) #time range
    
    #transient response for m
    transient_m = lambda m, t: n_derivative(m,t,alpha_m(V_c),beta_m(V_c))
    response_m = euler(transient_m,m0,t_max,dt)[0]
    
    #transient response for h
    transient_h = lambda h, t: n_derivative(h,t,alpha_h(V_c),beta_h(V_c))
    response_h = euler(transient_h,m0,t_max,dt)[0]
    
    #here we calculate the current responses I over time
    voltage = np.linspace(0,t_max*dt,t_max)
    for i in range(t_max): 
        voltage[i] = voltage_clamp(time[i],V_c,t_change)
    current_response = current_na(response_m,response_h,V_c,g_na,E_na)
    
    plt.plot(time,current_response,label='%i mV' %V_c)
```

```python
plt.figure(figsize=(10,6))
[plot_is_clamped(0.0529,0.5961,-65+10*i,2000,0.005) for i in range(7)]
plt.title(r'Sodium Current Responses $I_{Na}$ under 7 various Voltage-Clamp Conditions')
plt.ylabel(r'current response $I_{Na}$ (mA)')
plt.xlabel('time (ms)')
plt.legend()
plt.grid()
plt.show()
```

Mechanisms responsible for the sodium current rise and decay are the time constant and steady-state values. Higher time constant for a given voltage makes the response faster. The steady-state changes what direction the response is going.

### d) Plot the instantaneous and steady-state I-V curves for sodium channels

```python
def plot_iv_sodium(m,h,vspan,points,lbl): #instantaneous
    
    g_na = 120e-6 #conductance sodium (mS)
    E_na = 50 #reversal potential sodium (mV)

    voltage = np.linspace(-vspan,vspan,points) #empty voltage
    current = current_na(m,h,voltage,g_na,E_na) #current calculation
    
    plt.plot(voltage,current,label=lbl)
```

```python
def plot_iv_sodium_steady(vspan,points,lbl): #steady-state
    
    g_na = 120e-6 #conductance sodium (mS)
    E_na = 50 #reversal potential sodium (mV)
    
    voltage = np.linspace(-vspan,vspan,points) #empty voltage
    steady_state_m = np.zeros(points) #empty steady-state m
    steady_state_h = np.zeros(points) #empty steady-state h
    
    for i in range(points):
        
        am = alpha_m(voltage[i])
        bm = beta_m(voltage[i])
        ah = alpha_h(voltage[i])
        bh = beta_h(voltage[i])
        steady_state_m[i] = steady_activation(am,bm)
        steady_state_h[i] = steady_activation(ah,bh)
    
    #current calculation
    current = current_na(steady_state_m,steady_state_h,voltage,g_na,E_na) 
    
    plt.plot(voltage,current,label=lbl)
```

```python
plt.figure(figsize=(10,6))
plot_iv_sodium(0.0529,0.5961,150,300,'instantaneous')
plt.title('I-V Relation for Sodium Current')
plt.ylabel('current $I_{Na}$ (mA)')
plt.xlabel('voltage $V$ (mV)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plot_iv_sodium_steady(150,300,'steady-state')
plt.title('I-V Relation for Sodium Current')
plt.ylabel('current $I_{Na}$ (mA)')
plt.xlabel('voltage $V$ (mV)')
plt.legend()
plt.grid()
plt.show()
```

The instantaneous plots for both potassium and sodium are the same. However, for the steady-states the potassium is a sigmoid curve while the sodium is a dip down. The sodium steady-state may represent an inhibatory spike from the sodium, where it has a max inhibition at some voltage. For the positive IV curve of potassium, the model shows that a higher voltage returns a higher current. However, this model can become saturated which is an interesting advantage over previous observed models.

### (bonus) Investigate the sum of both currents

```python
def both_currents(m0,h0,n0,V,t_max,dt):
    
    t_change = 2 #time condition (ms)
    time = np.linspace(0,t_max*dt,t_max) #time range
    
    #stuff for potassium-----
    g_k = 36e-6 #conductance potassium (mS)
    E_k = -77 #reversal potential potassium (mV)
    
    #here we calculate the transient responses n over time
    n_derivative_clamp2 = lambda n, t: n_derivative_clamp(n,t,V,t_change)
    transient_response = euler(n_derivative_clamp2,n0,t_max,dt)[0]
    
    #here we calculate the current responses I over time
    voltage = np.linspace(0,t_max*dt,t_max)
    for i in range(t_max): 
        voltage[i] = voltage_clamp(time[i],V,t_change)
        
    current_potassium = current_k(transient_response,g_k,voltage,E_k)
    
    
    #stuff for sodium-----
    g_na = 120e-6 #conductance sodium (mS)
    E_na = 50 #reversal potential sodium (mV)
    
    #transient response for m
    transient_m = lambda m, t: n_derivative(m,t,alpha_m(V),beta_m(V))
    m = euler(transient_m,m0,t_max,dt)[0]
    
    #transient response for h
    transient_h = lambda h, t: n_derivative(h,t,alpha_h(V),beta_h(V))
    h = euler(transient_h,m0,t_max,dt)[0]
    
    current_sodium = current_na(m,h,V,g_na,E_na)
    
    total_current = -1 * (current_potassium + current_sodium)
    
    plt.plot(time,total_current,label='- %i mV' %V)
```

```python
plt.figure(figsize=(10,6))
[both_currents(0.0529,0.5961,0.3177,-65+20*i,2000,0.005) for i in range(4)]
plt.title(r'Total Current Responses $I_{K} + I_{Na}$ under 4 various Voltage-Clamp Conditions')
plt.ylabel(r'current response (mA)')
plt.xlabel('time (ms)')
plt.legend()
plt.grid()
plt.show()
```
