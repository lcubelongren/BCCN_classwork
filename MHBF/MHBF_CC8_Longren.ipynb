## #1. Good vs. bad Döner

### 1. The agent:

## <font color='k'>  $ ∆Q(s_t, a_t) = η(r_t + γQ(s_{t+1}, a_{t+1}) − Q(s_t, a_t)) $  ,  SARSA

```python
def epsilon_greedy(Q, state, epsilon):
    greedy_choice = np.random.choice([True, False], p=[1-epsilon, epsilon])
    if greedy_choice:  # hit the 1-epsilon choice
        if np.argmax(Q):  # down movement
            action = 'down'
        else:  # up movement
            action = 'up'
    else:  # hit the epsilon choice
        if np.random.randint(2):  # random choice
            action = 'down'
        else:
            action = 'up'
    return action
```

```python
def agent(state_num, trial_num, eta, epsilon, gamma): 
    'initialize the Q-values, and learn with SARSA'
    'following the epsilon greedy policy'
    rewards = np.full(state_num+2, -10)  # possible rewards for each state 
    rewards[1] = 1000  # bottom-most state
    rewards[-2] = 100  # top-most state
    rewards[0], rewards[-1] = 1000, 100
    Q = np.random.uniform(0, 5, (state_num+2, 2))  # initialize Q(s,a) (state, direction) 
    Q[0], Q[-1] = [0,0], [0,0]  # boundary conditons
    trial_count = []  # tracking duration
    for _ in range(trial_num):  # loop over episode
        state = np.random.randint(1, state_num)  # initial starting position
        action = epsilon_greedy(Q[state], state, epsilon)  # initial action
        count = 0  # track loop number for each trial
        while state != 0 and state != state_num:  # while not at a reward
            count += 1
            'choose the movement from the Q-values'
            if action == 'up':
                new_state = state + 1
            elif action == 'down':
                new_state = state - 1
            new_action = epsilon_greedy(Q[new_state], new_state, epsilon)  # choose new action
            'update the Q-value from the given reward'
            if action == 'up':
                Q[state][0] += eta * (rewards[state] + gamma*Q[new_state][0] - Q[state][0])
            elif action == 'down':
                Q[state][1] += eta * (rewards[state] + gamma*Q[new_state][1] - Q[state][1])  
            state = new_state
            action = new_action
        trial_count = np.append(trial_count, count)
    return Q, trial_count
```

### 2. Learn for 50,000 trials:

```python
trial_num = 50000
state_num = 50
eta, epsilon, gamma = 0.1, 0.1, 1
Q, trial_count = agent(state_num=state_num, trial_num=trial_num, eta=eta, epsilon=epsilon, gamma=gamma)
```

```python
trial_range = np.arange(0, len(trial_count), 1)
plt.plot(trial_range[:500], trial_count[:500], color='k')
plt.title('Trial Duration')
plt.ylabel('duration')
plt.xlabel('trial number')
plt.grid()
plt.show()
```

The behavior of the agent improves much faster than the Q-values. The reason for this is that the length of the state space is so short, and the agent already has an idea of where the optimal rewards are by the time the Q-values have had the time to adjust. The Q-values converge to the value of $\eta \ r_t$ assuming $\gamma$ is 1, which in our case is 100 for the optimal reward. With the negative rewards of -10 for all other states, Q takes much longer to improve than the duration of the trial.

```python
def _run_trial(self):
    """
    Run a single trial on the gridworld until the agent reaches the reward position.
    Return the time it takes to get there.
    """
    # Initialize the latency (time to reach the target) for this trial
    latency = 0

    # Choose a random initial position and make sure that it is not in the wall.
    self.x_position = np.random.randint(0, self.N)
    self.y_position = np.random.randint(0, self.N)
    while self._is_wall(x_position=self.x_position, y_position=self.y_position) == True:
        self.x_position = np.random.uniform(0, self.N)
        self.y_position = np.random.uniform(0, self.N)
    
    # Run the trial by choosing an action and repeatedly applying SARSA
    # until the reward has been reached.
    self._choose_action  # initial action
    while self._arrived == False:  # until at the reward
        latency += 1
        self._update_state  # new state
        self.choose_action  # new action 
        self._update_Q  # update Q with SARSA
        
    return latency
```
